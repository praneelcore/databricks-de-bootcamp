# Section 1: Linked Lists

## What is a Linked List?

A linked list is a sequence of **nodes**, where each node contains:
- A **value** (the data)
- A **pointer** to the next node

```
Array:         [10] [20] [30] [40] [50]
               Contiguous memory -- fast index access

Linked List:   [10|->] [20|->] [30|->] [40|->] [50|None]
               Scattered memory -- no index access, but fast insert/delete
```

## Array vs Linked List

| Operation | Array | Linked List |
|-----------|-------|-------------|
| Access by index | O(1) | O(n) |
| Insert at front | O(n) | O(1) |
| Insert at back | O(1) amortized | O(n) or O(1) with tail pointer |
| Insert in middle | O(n) | O(1) if you have the node |
| Delete | O(n) | O(1) if you have the node |
| Memory | Contiguous | Scattered (extra pointer overhead) |

## When to Use Linked Lists

- Frequent insertions/deletions at the front
- Unknown size (dynamic)
- Implementing stacks and queues
- When you don't need random access

````python
# Linked List implementation from scratch
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    def append(self, val):
        """Add to end. O(n) without tail pointer."""
        new_node = ListNode(val)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        self.size += 1

    def prepend(self, val):
        """Add to front. O(1)."""
        new_node = ListNode(val, self.head)
        self.head = new_node
        self.size += 1

    def delete(self, val):
        """Delete first occurrence of val. O(n)."""
        if not self.head:
            return
        if self.head.val == val:
            self.head = self.head.next
            self.size -= 1
            return
        current = self.head
        while current.next:
            if current.next.val == val:
                current.next = current.next.next
                self.size -= 1
                return
            current = current.next

    def to_list(self):
        """Convert to Python list for display."""
        result = []
        current = self.head
        while current:
            result.append(current.val)
            current = current.next
        return result

    def reverse(self):
        """Reverse in-place. O(n) time, O(1) space."""
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev


# Demo
ll = LinkedList()
for val in [1, 2, 3, 4, 5]:
    ll.append(val)

print("Linked List Operations:")
print(f"  Initial: {ll.to_list()}")

ll.prepend(0)
print(f"  After prepend(0): {ll.to_list()}")

ll.delete(3)
print(f"  After delete(3): {ll.to_list()}")

ll.reverse()
print(f"  After reverse(): {ll.to_list()}")
````

# Section 2: Fast and Slow Pointers

## The Technique

Use two pointers moving at different speeds through a linked list.
Fast moves 2 steps, slow moves 1 step.

**Applications:**
- Detect a cycle (Floyd's algorithm)
- Find the middle of a linked list
- Find the start of a cycle
- Check if linked list is a palindrome

````python
# Fast and slow pointer patterns
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def make_list(values, cycle_pos=-1):
    """Create linked list, optionally with a cycle."""
    if not values:
        return None
    nodes = [ListNode(v) for v in values]
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]
    if cycle_pos >= 0:
        nodes[-1].next = nodes[cycle_pos]
    return nodes[0]

# Pattern 1: Find middle of linked list
def find_middle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow.val

# Pattern 2: Detect cycle
def has_cycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# Pattern 3: Reverse a linked list
def reverse_list(head):
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev

print("FAST AND SLOW POINTER PATTERNS")
print("=" * 60)

# Test find middle
for values in [[1,2,3,4,5], [1,2,3,4], [1], [1,2]]:
    head = make_list(values)
    mid = find_middle(head)
    print(f"  Middle of {values}: {mid}")

# Test cycle detection
print()
no_cycle = make_list([1, 2, 3, 4, 5])
with_cycle = make_list([1, 2, 3, 4, 5], cycle_pos=2)
print(f"  [1,2,3,4,5] has cycle: {has_cycle(no_cycle)}")
print(f"  [1,2,3,4,5] with cycle at pos 2: {has_cycle(with_cycle)}")

# Test reverse
head = make_list([1, 2, 3, 4, 5])
reversed_head = reverse_list(head)
result = []
while reversed_head:
    result.append(reversed_head.val)
    reversed_head = reversed_head.next
print(f"\n  Reversed [1,2,3,4,5]: {result}")
````

# Section 3: Stacks

## Stack = Last In, First Out (LIFO)

```
Push 1: [1]
Push 2: [1, 2]
Push 3: [1, 2, 3]
Pop:    [1, 2]    <- 3 removed (last in, first out)
Pop:    [1]       <- 2 removed
```

## Stack Applications

| Application | How Stack Helps |
|-------------|----------------|
| Undo/Redo | Push actions, pop to undo |
| Function call stack | Each function call pushed, popped on return |
| Bracket matching | Push open brackets, pop on close |
| Expression evaluation | Push operands, pop when operator seen |
| DFS traversal | Push neighbors, pop to visit |
| Monotonic stack | Maintain sorted order for "next greater element" |

## Monotonic Stack -- A Powerful Pattern

A monotonic stack maintains elements in sorted order.
Used for "next greater/smaller element" problems.

```
Find next greater element for each position:
Array: [2, 1, 5, 3, 6, 4]

Process 2: stack=[2]
Process 1: stack=[2,1]  (1 < 2, push)
Process 5: 5 > 1, pop 1 (next greater of 1 is 5)
           5 > 2, pop 2 (next greater of 2 is 5)
           stack=[5]
...
```

````python
# Stack patterns
print("STACK PATTERNS")
print("=" * 60)

# Pattern 1: Valid parentheses
def is_valid(s):
    stack = []
    pairs = {")": "(", "}": "{", "]": "["}
    for char in s:
        if char in "({[":
            stack.append(char)
        elif char in ")}]":
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()
    return len(stack) == 0

print("\n1. Valid Parentheses:")
for s in ["()", "()[]{}", "(]", "([)]", "{[]}"]:
    print(f"   '{s}': {is_valid(s)}")

# Pattern 2: Monotonic stack -- Next Greater Element
def next_greater_element(nums):
    result = [-1] * len(nums)
    stack = []  # Stack of indices
    for i, num in enumerate(nums):
        while stack and nums[stack[-1]] < num:
            idx = stack.pop()
            result[idx] = num
        stack.append(i)
    return result

print("\n2. Next Greater Element:")
for nums in [[2, 1, 5, 3, 6, 4], [1, 2, 3, 4], [4, 3, 2, 1]]:
    print(f"   {nums} -> {next_greater_element(nums)}")

# Pattern 3: Daily temperatures (monotonic stack)
def daily_temperatures(temps):
    result = [0] * len(temps)
    stack = []  # Stack of indices
    for i, temp in enumerate(temps):
        while stack and temps[stack[-1]] < temp:
            idx = stack.pop()
            result[idx] = i - idx  # Days to wait
        stack.append(i)
    return result

temps = [73, 74, 75, 71, 69, 72, 76, 73]
print(f"\n3. Daily Temperatures:")
print(f"   Temps: {temps}")
print(f"   Wait:  {daily_temperatures(temps)}")
print(f"   (73 waits 1 day for 74, 74 waits 1 day for 75, etc.)")
````

# Section 4: Queues and Deques

## Queue = First In, First Out (FIFO)

```
Enqueue 1: [1]
Enqueue 2: [1, 2]
Enqueue 3: [1, 2, 3]
Dequeue:   [2, 3]    <- 1 removed (first in, first out)
Dequeue:   [3]       <- 2 removed
```

## Deque = Double-Ended Queue

```python
from collections import deque
d = deque()
d.append(1)       # Add to right: O(1)
d.appendleft(0)   # Add to left: O(1)
d.pop()           # Remove from right: O(1)
d.popleft()       # Remove from left: O(1)
```

## Queue Applications

- BFS (Breadth-First Search)
- Task scheduling
- Sliding window maximum (monotonic deque)
- Level-order tree traversal

````python
# Queue and deque patterns
from collections import deque

print("QUEUE AND DEQUE PATTERNS")
print("=" * 60)

# Pattern 1: BFS using queue
def bfs_grid(grid, start, end):
    """Find shortest path in grid using BFS."""
    rows, cols = len(grid), len(grid[0])
    queue = deque([(start, 0)])  # (position, distance)
    visited = {start}

    while queue:
        (r, c), dist = queue.popleft()
        if (r, c) == end:
            return dist
        for dr, dc in [(0,1),(0,-1),(1,0),(-1,0)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols and (nr,nc) not in visited and grid[nr][nc] != 1:
                visited.add((nr, nc))
                queue.append(((nr, nc), dist + 1))
    return -1

# 0 = open, 1 = wall
grid = [
    [0, 0, 0, 0],
    [1, 1, 0, 1],
    [0, 0, 0, 0],
    [0, 1, 1, 0],
]
print(f"\n1. BFS Shortest Path:")
print(f"   Grid: 4x4 with walls")
dist = bfs_grid(grid, (0,0), (3,3))
print(f"   Shortest path from (0,0) to (3,3): {dist} steps")

# Pattern 2: Sliding window maximum (monotonic deque)
def sliding_window_max(nums, k):
    """Find max in each window of size k. O(n) time."""
    result = []
    dq = deque()  # Stores indices, decreasing values

    for i, num in enumerate(nums):
        # Remove elements outside window
        while dq and dq[0] < i - k + 1:
            dq.popleft()
        # Remove smaller elements (they can never be max)
        while dq and nums[dq[-1]] < num:
            dq.pop()
        dq.append(i)
        # Window is full
        if i >= k - 1:
            result.append(nums[dq[0]])

    return result

nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
print(f"\n2. Sliding Window Maximum:")
print(f"   nums={nums}, k={k}")
print(f"   Max in each window: {sliding_window_max(nums, k)}")
````

# Summary

## Key Takeaways

1. **Linked List**: O(1) insert/delete at known position, O(n) access by index
2. **Fast/Slow Pointers**: Detect cycles, find middle, O(1) space
3. **Stack (LIFO)**: Undo, bracket matching, DFS, monotonic stack
4. **Queue (FIFO)**: BFS, task scheduling, level-order traversal
5. **Monotonic Stack**: O(n) solution for "next greater/smaller element" problems
6. **Deque**: O(1) operations at both ends -- use for sliding window maximum
