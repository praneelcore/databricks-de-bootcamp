# Section 1: Recursion Fundamentals

## What is Recursion?

A function that calls itself. Every recursive solution has:
1. **Base case**: When to stop (prevents infinite recursion)
2. **Recursive case**: How to break the problem into smaller subproblems

```python
def factorial(n):
    # Base case
    if n <= 1:
        return 1
    # Recursive case: n! = n * (n-1)!
    return n * factorial(n - 1)
```

## The Call Stack

```
factorial(4)
  -> 4 * factorial(3)
       -> 3 * factorial(2)
            -> 2 * factorial(1)
                 -> 1  (base case!)
            <- 2 * 1 = 2
       <- 3 * 2 = 6
  <- 4 * 6 = 24
```

## Recursion vs Iteration

| Aspect | Recursion | Iteration |
|--------|-----------|-----------|
| Readability | Often cleaner | Can be verbose |
| Stack space | O(depth) -- can overflow | O(1) |
| Performance | Overhead per call | Faster |
| Use when | Tree/graph traversal, divide & conquer | Simple loops |

````python
# Recursion patterns
import sys
sys.setrecursionlimit(10000)

print("RECURSION PATTERNS")
print("=" * 60)

# Pattern 1: Fibonacci (naive -- exponential)
def fib_naive(n):
    if n <= 1:
        return n
    return fib_naive(n-1) + fib_naive(n-2)

# Pattern 2: Fibonacci (memoized -- linear)
def fib_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

import time
print("\n1. Fibonacci -- Naive vs Memoized:")
for n in [10, 20, 30]:
    start = time.perf_counter()
    result_naive = fib_naive(n)
    t_naive = time.perf_counter() - start

    start = time.perf_counter()
    result_memo = fib_memo(n, {})
    t_memo = time.perf_counter() - start

    print(f"   fib({n}) = {result_naive}")
    print(f"     Naive: {t_naive:.6f}s | Memoized: {t_memo:.8f}s")

# Pattern 3: Power function (divide and conquer)
def power(base, exp):
    """x^n in O(log n) time using divide and conquer."""
    if exp == 0:
        return 1
    if exp % 2 == 0:
        half = power(base, exp // 2)
        return half * half
    return base * power(base, exp - 1)

print(f"\n2. Fast Power (O(log n)):")
print(f"   2^10 = {power(2, 10)}")
print(f"   3^5 = {power(3, 5)}")
````

# Section 2: Backtracking

## The Framework

Backtracking = DFS + undo

```python
def backtrack(state, choices):
    if is_solution(state):
        add_to_results(state)
        return

    for choice in choices:
        if is_valid(choice, state):
            make_choice(state, choice)    # Choose
            backtrack(state, choices)     # Explore
            undo_choice(state, choice)    # Unchoose (backtrack!)
```

## When to Use Backtracking

- Generate all permutations/combinations/subsets
- Constraint satisfaction (Sudoku, N-Queens)
- Path finding with constraints
- Word search in a grid

````python
# Backtracking patterns
print("BACKTRACKING PATTERNS")
print("=" * 60)

# Pattern 1: All permutations
def permutations(nums):
    result = []
    def backtrack(current, remaining):
        if not remaining:
            result.append(current[:])
            return
        for i in range(len(remaining)):
            current.append(remaining[i])
            backtrack(current, remaining[:i] + remaining[i+1:])
            current.pop()
    backtrack([], nums)
    return result

print(f"\n1. Permutations of [1,2,3]:")
perms = permutations([1, 2, 3])
print(f"   Count: {len(perms)}")
print(f"   First 4: {perms[:4]}")

# Pattern 2: All subsets
def subsets(nums):
    result = []
    def backtrack(start, current):
        result.append(current[:])
        for i in range(start, len(nums)):
            current.append(nums[i])
            backtrack(i + 1, current)
            current.pop()
    backtrack(0, [])
    return result

print(f"\n2. Subsets of [1,2,3]:")
subs = subsets([1, 2, 3])
print(f"   Count: {len(subs)} (2^3 = 8)")
print(f"   All: {subs}")

# Pattern 3: Combination sum
def combination_sum(candidates, target):
    result = []
    def backtrack(start, current, remaining):
        if remaining == 0:
            result.append(current[:])
            return
        for i in range(start, len(candidates)):
            if candidates[i] <= remaining:
                current.append(candidates[i])
                backtrack(i, current, remaining - candidates[i])
                current.pop()
    backtrack(0, [], target)
    return result

print(f"\n3. Combination Sum (target=7, candidates=[2,3,6,7]):")
result = combination_sum([2, 3, 6, 7], 7)
print(f"   Result: {result}")
````

# Section 3: Dynamic Programming

## What is DP?

DP solves problems by breaking them into **overlapping subproblems**
and storing results to avoid recomputation.

**Two conditions for DP:**
1. **Optimal substructure**: Optimal solution contains optimal solutions to subproblems
2. **Overlapping subproblems**: Same subproblems solved multiple times

## Two Approaches

```
TOP-DOWN (Memoization):          BOTTOM-UP (Tabulation):
Start with the big problem       Start with smallest subproblems
Recurse down                     Build up to the answer
Cache results                    Fill a table

def fib(n, memo={}):             def fib(n):
    if n in memo: return memo[n]     dp = [0] * (n+1)
    if n <= 1: return n              dp[1] = 1
    memo[n] = fib(n-1) + fib(n-2)   for i in range(2, n+1):
    return memo[n]                       dp[i] = dp[i-1] + dp[i-2]
                                     return dp[n]
```

## The DP Framework

1. **Define the state**: What does dp[i] represent?
2. **Base case**: What are the smallest subproblems?
3. **Transition**: How does dp[i] relate to dp[i-1], dp[i-2], etc.?
4. **Answer**: Which dp value is the final answer?

````python
# Dynamic Programming patterns
print("DYNAMIC PROGRAMMING PATTERNS")
print("=" * 60)

# Pattern 1: Coin Change (minimum coins)
def coin_change(coins, amount):
    """Minimum coins to make amount. O(amount * len(coins)) time."""
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins to make amount 0

    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float("inf") else -1

print("\n1. Coin Change (minimum coins):")
tests = [([1,5,11], 15), ([1,2,5], 11), ([2], 3)]
for coins, amount in tests:
    result = coin_change(coins, amount)
    print(f"   coins={coins}, amount={amount} -> {result} coins")

# Pattern 2: Longest Common Subsequence
def lcs(s1, s2):
    """LCS length. O(m*n) time and space."""
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

print("\n2. Longest Common Subsequence:")
pairs = [("abcde", "ace"), ("abc", "abc"), ("abc", "def")]
for s1, s2 in pairs:
    print(f"   LCS('{s1}', '{s2}') = {lcs(s1, s2)}")

# Pattern 3: 0/1 Knapsack
def knapsack(weights, values, capacity):
    """Max value with weight limit. O(n*capacity) time."""
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(capacity + 1):
            dp[i][w] = dp[i-1][w]  # Don't take item i
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i][w], dp[i-1][w-weights[i-1]] + values[i-1])

    return dp[n][capacity]

weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
capacity = 8
print(f"\n3. 0/1 Knapsack:")
print(f"   weights={weights}, values={values}, capacity={capacity}")
print(f"   Max value: {knapsack(weights, values, capacity)}")

# Pattern 4: Climbing Stairs
def climb_stairs(n):
    """Ways to climb n stairs (1 or 2 steps at a time)."""
    if n <= 2:
        return n
    dp = [0] * (n + 1)
    dp[1], dp[2] = 1, 2
    for i in range(3, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

print(f"\n4. Climbing Stairs:")
for n in range(1, 8):
    print(f"   {n} stairs: {climb_stairs(n)} ways")
````

# Summary

## Key Takeaways

1. **Recursion**: Base case + recursive case. Watch for stack overflow.
2. **Backtracking**: Choose -> Explore -> Unchoose. For generating all possibilities.
3. **DP**: Cache overlapping subproblems. Define state, base case, transition.
4. **Memoization (top-down)**: Easier to write, same complexity as tabulation
5. **Tabulation (bottom-up)**: No recursion overhead, sometimes easier to optimize space
