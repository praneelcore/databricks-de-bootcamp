# Section 1: Binary Trees

## Structure

```
        1          <- Root (level 0)
       / \
      2   3        <- Level 1
     / \   \
    4   5   6      <- Level 2 (leaves)
```

Each node has at most 2 children (left and right).

## Key Terms

| Term | Definition |
|------|-----------|
| **Root** | Top node (no parent) |
| **Leaf** | Node with no children |
| **Height** | Longest path from root to leaf |
| **Depth** | Distance from root to a node |
| **Balanced** | Height difference between subtrees <= 1 |
| **Complete** | All levels full except possibly last |
| **BST** | Left < Node < Right for all nodes |

## The Four Traversals

```
Tree:     1
         / \
        2   3
       / \
      4   5

Inorder   (Left, Root, Right): 4, 2, 5, 1, 3  <- Sorted for BST!
Preorder  (Root, Left, Right): 1, 2, 4, 5, 3  <- Good for copying tree
Postorder (Left, Right, Root): 4, 5, 2, 3, 1  <- Good for deleting tree
BFS/Level-order:               1, 2, 3, 4, 5  <- Level by level
```

````python
# Binary Tree implementation and traversals
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def make_tree(values):
    """Build tree from level-order list (None = missing node)."""
    if not values:
        return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

# Traversals
def inorder(root):
    if not root:
        return []
    return inorder(root.left) + [root.val] + inorder(root.right)

def preorder(root):
    if not root:
        return []
    return [root.val] + preorder(root.left) + preorder(root.right)

def postorder(root):
    if not root:
        return []
    return postorder(root.left) + postorder(root.right) + [root.val]

def bfs(root):
    if not root:
        return []
    result, queue = [], deque([root])
    while queue:
        node = queue.popleft()
        result.append(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return result

# Build tree: [1, 2, 3, 4, 5, None, 6]
root = make_tree([1, 2, 3, 4, 5, None, 6])

print("BINARY TREE TRAVERSALS")
print("=" * 60)
print(f"Tree: [1, 2, 3, 4, 5, None, 6]")
print(f"  Inorder   (L,Root,R): {inorder(root)}")
print(f"  Preorder  (Root,L,R): {preorder(root)}")
print(f"  Postorder (L,R,Root): {postorder(root)}")
print(f"  BFS (level-order):    {bfs(root)}")
````

# Section 2: Common Tree Problems

## Height, Diameter, Path Sum, LCA

These are the most common tree interview problems.

````python
# Common tree problems
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def make_tree(values):
    from collections import deque
    if not values:
        return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

# 1. Maximum depth (height)
def max_depth(root):
    if not root:
        return 0
    return 1 + max(max_depth(root.left), max_depth(root.right))

# 2. Check if balanced
def is_balanced(root):
    def height(node):
        if not node:
            return 0
        left_h = height(node.left)
        if left_h == -1:
            return -1
        right_h = height(node.right)
        if right_h == -1:
            return -1
        if abs(left_h - right_h) > 1:
            return -1
        return 1 + max(left_h, right_h)
    return height(root) != -1

# 3. Path sum (does a root-to-leaf path sum to target?)
def has_path_sum(root, target):
    if not root:
        return False
    if not root.left and not root.right:  # Leaf
        return root.val == target
    return (has_path_sum(root.left, target - root.val) or
            has_path_sum(root.right, target - root.val))

# 4. Lowest Common Ancestor
def lowest_common_ancestor(root, p, q):
    if not root or root.val == p or root.val == q:
        return root
    left = lowest_common_ancestor(root.left, p, q)
    right = lowest_common_ancestor(root.right, p, q)
    if left and right:
        return root  # p and q are in different subtrees
    return left or right

print("COMMON TREE PROBLEMS")
print("=" * 60)

tree = make_tree([3, 9, 20, None, None, 15, 7])
print(f"\nTree: [3, 9, 20, None, None, 15, 7]")
print(f"  Max depth: {max_depth(tree)}")
print(f"  Is balanced: {is_balanced(tree)}")

tree2 = make_tree([5, 4, 8, 11, None, 13, 4, 7, 2, None, None, None, 1])
print(f"\nTree2: [5, 4, 8, 11, None, 13, 4, 7, 2, None, None, None, 1]")
print(f"  Has path sum 22: {has_path_sum(tree2, 22)}")
print(f"  Has path sum 26: {has_path_sum(tree2, 26)}")

tree3 = make_tree([3, 5, 1, 6, 2, 0, 8, None, None, 7, 4])
lca = lowest_common_ancestor(tree3, 5, 1)
print(f"\nTree3: LCA of 5 and 1: {lca.val}")
lca2 = lowest_common_ancestor(tree3, 5, 4)
print(f"  LCA of 5 and 4: {lca2.val}")
````

# Section 3: Heaps (Priority Queues)

## What is a Heap?

A heap is a **complete binary tree** where every parent is smaller (min-heap)
or larger (max-heap) than its children.

```
Min-Heap:          Max-Heap:
     1                  9
    / \                / \
   3   2              7   8
  / \ / \            / \ / \
 7  4 5  6          2  4 3  1

Parent <= Children    Parent >= Children
```

## Python heapq Module

Python only has min-heap. For max-heap, negate values.

```python
import heapq

# Min-heap
heap = [5, 3, 8, 1, 9]
heapq.heapify(heap)        # O(n) -- convert list to heap
heapq.heappush(heap, 2)    # O(log n) -- add element
min_val = heapq.heappop(heap)  # O(log n) -- remove minimum
min_val = heap[0]          # O(1) -- peek minimum (don't remove)

# Max-heap (negate values)
max_heap = [-x for x in [5, 3, 8, 1, 9]]
heapq.heapify(max_heap)
max_val = -heapq.heappop(max_heap)  # Negate back

# Top-K elements
top_3 = heapq.nlargest(3, [5, 3, 8, 1, 9])   # [9, 8, 5]
bot_3 = heapq.nsmallest(3, [5, 3, 8, 1, 9])  # [1, 3, 5]
```

````python
# Heap patterns
import heapq

print("HEAP (PRIORITY QUEUE) PATTERNS")
print("=" * 60)

# Pattern 1: Kth largest element
def find_kth_largest(nums, k):
    """Use min-heap of size k. O(n log k) time."""
    heap = []
    for num in nums:
        heapq.heappush(heap, num)
        if len(heap) > k:
            heapq.heappop(heap)  # Remove smallest
    return heap[0]  # Smallest in heap = kth largest overall

nums = [3, 2, 1, 5, 6, 4]
print(f"\n1. Kth Largest:")
for k in [1, 2, 3]:
    print(f"   {k}th largest in {nums}: {find_kth_largest(nums, k)}")

# Pattern 2: Top K frequent elements
def top_k_frequent(nums, k):
    """O(n log k) time."""
    from collections import Counter
    freq = Counter(nums)
    return heapq.nlargest(k, freq.keys(), key=freq.get)

nums = [1, 1, 1, 2, 2, 3]
print(f"\n2. Top K Frequent:")
print(f"   nums={nums}")
print(f"   Top 2 frequent: {top_k_frequent(nums, 2)}")

# Pattern 3: Merge K sorted lists
def merge_k_sorted(lists):
    """Merge k sorted lists. O(n log k) time."""
    heap = []
    for i, lst in enumerate(lists):
        if lst:
            heapq.heappush(heap, (lst[0], i, 0))

    result = []
    while heap:
        val, list_idx, elem_idx = heapq.heappop(heap)
        result.append(val)
        if elem_idx + 1 < len(lists[list_idx]):
            next_val = lists[list_idx][elem_idx + 1]
            heapq.heappush(heap, (next_val, list_idx, elem_idx + 1))

    return result

lists = [[1, 4, 7], [2, 5, 8], [3, 6, 9]]
print(f"\n3. Merge K Sorted Lists:")
print(f"   Input: {lists}")
print(f"   Merged: {merge_k_sorted(lists)}")

# Pattern 4: Find median from data stream
class MedianFinder:
    def __init__(self):
        self.small = []  # Max-heap (negated) for lower half
        self.large = []  # Min-heap for upper half

    def add_num(self, num):
        heapq.heappush(self.small, -num)
        # Balance: move largest of small to large
        heapq.heappush(self.large, -heapq.heappop(self.small))
        # Keep small >= large in size
        if len(self.large) > len(self.small):
            heapq.heappush(self.small, -heapq.heappop(self.large))

    def find_median(self):
        if len(self.small) > len(self.large):
            return -self.small[0]
        return (-self.small[0] + self.large[0]) / 2

mf = MedianFinder()
print(f"\n4. Median from Data Stream:")
for num in [1, 2, 3, 4, 5]:
    mf.add_num(num)
    print(f"   After adding {num}: median = {mf.find_median()}")
````

# Summary

## Key Takeaways

1. **Tree traversals**: Inorder (sorted for BST), Preorder (copy), Postorder (delete), BFS (level-order)
2. **Recursion is natural for trees**: Most tree problems = base case (null) + recursive calls on left/right
3. **Height/depth**: Use DFS recursion, return 0 for null
4. **Heap**: O(log n) insert/delete, O(1) peek min/max
5. **Top-K pattern**: Use min-heap of size K -- O(n log k) vs O(n log n) sort
