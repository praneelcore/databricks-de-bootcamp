# 📗 Section 1: Cloud Services Comparison

## Storage Services

| Service | AWS | Azure | GCP |
|---------|-----|-------|-----|
| **Object Storage** | S3 | ADLS Gen2 | GCS |
| **Cost (per GB/month)** | $0.023 | $0.018 | $0.020 |
| **Max object size** | 5 TB | 190.7 TB | 5 TB |
| **Consistency** | Strong | Strong | Strong |
| **Access patterns** | Bucket/key | Container/blob | Bucket/object |

## Compute Services

| Service | AWS | Azure | GCP |
|---------|-----|-------|-----|
| **Managed Spark** | EMR | HDInsight | Dataproc |
| **Serverless ETL** | Glue | Data Factory | Dataflow |
| **Streaming** | Kinesis | Event Hubs | Pub/Sub |
| **Orchestration** | Step Functions, MWAA | Data Factory | Cloud Composer |
| **Data Warehouse** | Redshift | Synapse | BigQuery |
| **Databricks** | ✅ Available | ✅ Available | ✅ Available |

## Databricks on Each Cloud

| Aspect | AWS | Azure | GCP |
|--------|-----|-------|-----|
| **Storage** | S3 | ADLS Gen2 | GCS |
| **Identity** | IAM roles | Azure AD / Entra ID | IAM |
| **Networking** | VPC | VNet | VPC |
| **Secrets** | AWS Secrets Manager | Key Vault | Secret Manager |
| **Catalog** | Glue Catalog (legacy) | Unity Catalog | Unity Catalog |

## Multi-Cloud Strategy

```
When to go multi-cloud:
✅ Regulatory requirements (data residency)
✅ Acquisitions (inherited different clouds)
✅ Avoid vendor lock-in (strategic decision)
✅ Best-of-breed (BigQuery for analytics + AWS for ML)

When to stay single-cloud:
✅ Simpler operations (one set of skills)
✅ Lower cost (no cross-cloud data transfer)
✅ Faster development (one ecosystem to learn)
✅ Better integration (native services work together)
```

````python
# Cloud cost comparison calculator
def compare_cloud_costs(data_size_tb, compute_hours_month, queries_per_day):
    """Compare costs across AWS, Azure, and GCP for a data platform."""
    
    costs = {
        "AWS": {
            "storage": data_size_tb * 1024 * 0.023,  # S3 Standard
            "compute": compute_hours_month * 0.384 * 1.3,  # EMR + EC2
            "queries": queries_per_day * 30 * 0.005,  # Athena per query
            "databricks": compute_hours_month * 0.55 * 10,  # DBU cost
        },
        "Azure": {
            "storage": data_size_tb * 1024 * 0.018,  # ADLS Gen2
            "compute": compute_hours_month * 0.384 * 1.2,  # VMs
            "queries": queries_per_day * 30 * 0.004,  # Synapse serverless
            "databricks": compute_hours_month * 0.55 * 10,
        },
        "GCP": {
            "storage": data_size_tb * 1024 * 0.020,  # GCS
            "compute": compute_hours_month * 0.384 * 1.1,  # GCE
            "queries": queries_per_day * 30 * 5 * 0.005,  # BigQuery per TB scanned
            "databricks": compute_hours_month * 0.55 * 10,
        }
    }
    
    print(f"💰 CLOUD COST COMPARISON")
    print(f"   Data: {data_size_tb} TB | Compute: {compute_hours_month} hrs/mo | Queries: {queries_per_day}/day")
    print("=" * 60)
    print(f"{'Component':<15} {'AWS':<15} {'Azure':<15} {'GCP':<15}")
    print("-" * 60)
    
    for component in ["storage", "compute", "queries", "databricks"]:
        aws = costs["AWS"][component]
        azure = costs["Azure"][component]
        gcp = costs["GCP"][component]
        print(f"{component:<15} ${aws:>10,.0f}   ${azure:>10,.0f}   ${gcp:>10,.0f}")
    
    print("-" * 60)
    for cloud in ["AWS", "Azure", "GCP"]:
        total = sum(costs[cloud].values())
        print(f"{'TOTAL ' + cloud:<15} ${total:>10,.0f}")
    
    print("\n💡 Note: Actual costs vary significantly based on:")
    print("   • Region, reserved instances, spot pricing")
    print("   • Data transfer (cross-region/cross-cloud)")
    print("   • Specific service tiers and configurations")

compare_cloud_costs(data_size_tb=10, compute_hours_month=500, queries_per_day=200)
````

## Multi-Cloud Data Transfer

```

  Use Google Transfer Service (free for GCS ingress)
  Cost: AWS egress ($0.09/GB)

Recommendation: Avoid cross-cloud transfers where possible
  -> Keep compute and storage in same cloud/region
```
