# Section 3: databricks-tools-core -- The Python Library

## Overview

`databricks-tools-core` is a pure Python library providing clean, high-level APIs
for common Databricks operations. Install it with:

```
pip install databricks-tools-core
```

## Module Structure

```
databricks_tools_core/
├── auth.py                    # Authentication (contextvars + env vars)
├── sql/
│   ├── sql.py                 # execute_sql, execute_sql_multi
│   ├── warehouse.py           # list_warehouses, get_best_warehouse
│   └── table_stats.py         # get_table_stats_and_schema
├── jobs/
│   ├── jobs.py                # create_job, get_job, list_jobs, update_job, delete_job
│   └── runs.py                # run_job_now, get_run, wait_for_run, cancel_run
├── unity_catalog/
│   ├── catalogs.py            # list_catalogs, create_catalog
│   ├── schemas.py             # list_schemas, create_schema
│   └── tables.py              # list_tables, create_table, get_table
└── spark_declarative_pipelines/  # DLT/SDP pipeline management
```

## Authentication Priority

1. Context variables (for multi-user apps) -- set via `set_databricks_auth()`
2. Environment variables -- `DATABRICKS_HOST` and `DATABRICKS_TOKEN`
3. Config profile -- `DATABRICKS_CONFIG_PROFILE` or `~/.databrickscfg`

````python
# databricks-tools-core API reference
print("databricks-tools-core API Reference")
print("=" * 60)

api_reference = {
    "SQL Module (databricks_tools_core.sql)": {
        "execute_sql(sql_query, warehouse_id=None, catalog=None, schema=None)":
            "Execute a SQL query, returns list of dicts",
        "execute_sql_multi(sql_content, catalog=None, schema=None)":
            "Execute multiple SQL statements with dependency analysis (parallel where possible)",
        "list_warehouses(limit=20)":
            "List SQL warehouses (running ones first)",
        "get_best_warehouse()":
            "Auto-select best available warehouse",
        "get_table_stats_and_schema(catalog, schema, table_names=None, level=SIMPLE)":
            "Get table schemas and statistics (NONE/SIMPLE/DETAILED)",
    },
    "Jobs Module (databricks_tools_core.jobs)": {
        "create_job(name, tasks, schedule=None)":
            "Create a new job (serverless by default)",
        "list_jobs(name_filter=None, limit=20)":
            "List jobs with optional name filter",
        "run_job_now(job_id, notebook_params=None)":
            "Trigger a job run, returns run_id",
        "wait_for_run(run_id, timeout=3600)":
            "Wait for run completion, returns JobRunResult",
        "cancel_run(run_id)":
            "Cancel a running job",
    },
    "Unity Catalog Module (databricks_tools_core.unity_catalog)": {
        "catalogs.list_catalogs()":
            "List all accessible catalogs",
        "schemas.create_schema(catalog_name, schema_name, comment=None)":
            "Create a new schema",
        "tables.list_tables(catalog_name, schema_name)":
            "List tables in a schema",
        "tables.create_table(catalog, schema, table, columns, table_type)":
            "Create a new table",
    },
}

for module, functions in api_reference.items():
    print(f"\n{module}:")
    for func, description in functions.items():
        print(f"  {func}")
        print(f"    -> {description}")
````

````python
# Practical example: pipeline health check script using tools-core
# (Runs outside Databricks, connects to your workspace)

print("Example: Automated Pipeline Health Check Script")
print("=" * 60)

script_example = """
import os
from databricks_tools_core.sql import execute_sql, get_table_stats_and_schema
from databricks_tools_core.jobs import list_jobs, list_runs

# Authenticate via environment variables
os.environ["DATABRICKS_HOST"] = "https://your-workspace.cloud.databricks.com"
os.environ["DATABRICKS_TOKEN"] = "your-personal-access-token"

# 1. Check data freshness
freshness = execute_sql(
    "SELECT table_name, MAX(_ingested_at) AS last_updated, "
    "DATEDIFF(HOUR, MAX(_ingested_at), NOW()) AS hours_old "
    "FROM (SELECT 'orders' AS table_name, _ingested_at FROM prod.silver.orders "
    "UNION ALL SELECT 'customers', _ingested_at FROM prod.silver.customers) "
    "GROUP BY table_name"
)
for row in freshness:
    status = "STALE" if row["hours_old"] > 2 else "FRESH"
    print(f"  {row['table_name']}: {status} ({row['hours_old']}h old)")

# 2. Check recent job runs
jobs = list_jobs(name_filter="daily_pipeline")
for job in jobs:
    runs = list_runs(job_id=job["job_id"], limit=5)
    failed = [r for r in runs if r["state"]["result_state"] == "FAILED"]
    print(f"  {job['settings']['name']}: {len(failed)} failures in last 5 runs")

# 3. Get table statistics
stats = get_table_stats_and_schema(
    catalog="prod", schema="silver",
    table_names=["orders", "customers"],
)
for table in stats:
    print(f"  {table['name']}: {table['row_count']:,} rows")
"""

print(script_example)
print("This script runs OUTSIDE Databricks (on your laptop or in CI/CD)")
print("It connects to your workspace via the Databricks SDK")
````

# Section 5: MCP Server -- 50+ Tools for AI Assistants

## What is MCP?

Model Context Protocol (MCP) is an open standard that lets AI assistants call
external tools. The Databricks MCP server exposes 50+ Databricks operations
as callable tools.

```
AI Assistant (Claude Code / Cursor)
         |
         | "List all tables in prod.sales"
         v
MCP Server (databricks-mcp-server)
         |
         | Calls: list_tables(catalog="prod", schema="sales")
         v
Databricks Workspace
         |
         | Returns: [{"name": "orders", "rows": 20000}, ...]
         v
AI: "Found 5 tables: orders (20K rows), customers (5K rows)..."
```

## MCP Configuration

The installer creates this automatically in `.claude/mcp.json`:

```json
{
  "mcpServers": {
    "databricks": {
      "command": "uv",
      "args": ["run", "--project", "/path/to/ai-dev-kit", "databricks-mcp-server"],
      "env": {
        "DATABRICKS_HOST": "https://your-workspace.cloud.databricks.com",

}
```

# Section 7: AI Functions -- SQL-Native AI

## Built-in AI Functions

Databricks provides AI functions that run directly in SQL -- no Python needed.
These are part of the `databricks-ai-functions` skill.

Available functions:
- `ai_classify(text, labels)` -- Classify text into categories
- `ai_extract(text, fields)` -- Extract structured data from text
- `ai_summarize(text, max_words)` -- Summarize long text
- `ai_query(model, prompt)` -- Query any LLM with custom prompt
- `ai_forecast(date_col, value_col, horizon, frequency)` -- Time series forecasting
- `ai_parse_document(content)` -- Parse PDFs and images

## When to Use AI Functions

| Function | Use Case | Example |
|----------|----------|---------|
