## Architecture

```
INGESTION
  Lakeflow Connect (PostgreSQL CDC) --> Bronze

@dlt.table(name="bronze_orders")
def bronze_orders():
    return (spark.readStream
        .format("cloudFiles")
        .option("cloudFiles.format", "json")
        .option("cloudFiles.schemaLocation", "/checkpoints/orders/schema")
        .load("/Volumes/prod/raw/orders/"))
```

# dbt model for Gold layer
# models/marts/fct_daily_revenue.sql
SELECT
    order_date,
    COUNT(*) AS total_orders,

FROM {{ ref('silver_orders') }}
WHERE status = 'completed'
GROUP BY order_date

# Section 4: Cost Optimization for the Platform

## Monthly Cost Breakdown (Example: 20-person team)

| Component | Configuration | Monthly Cost |
|-----------|--------------|-------------|

| Dev clusters | Serverless notebooks | ~$2K |
| Storage (50TB Delta) | S3 Standard + lifecycle | ~$5K |
| **Total** | | **~$35K/month** |

## Top Cost Optimizations

1. **Serverless everywhere** -- eliminates idle compute cost
2. **Spot instances for batch** -- 60-80% savings on fault-tolerant jobs

4. **Liquid Clustering** -- reduces data scanned per query
5. **Predictive Optimization** -- automatic OPTIMIZE/VACUUM
6. **S3 lifecycle policies** -- move cold data to cheaper tiers
