# Section 1: Arrays -- The Foundation

## What is an Array?

An array (Python `list`) stores elements in **contiguous memory**, accessible by index.

```
Index:  0    1    2    3    4
Value: [10,  20,  30,  40,  50]
        ^                   ^
     arr[0]              arr[4]
```

## Time Complexities

| Operation | Time | Why |
|-----------|------|-----|
| Access by index `arr[i]` | O(1) | Direct memory address calculation |
| Append to end `arr.append(x)` | O(1) amortized | Occasional resize |
| Insert at position `arr.insert(i, x)` | O(n) | Must shift elements right |
| Delete at position `del arr[i]` | O(n) | Must shift elements left |
| Search `x in arr` | O(n) | Must check each element |
| Sort `arr.sort()` | O(n log n) | Timsort |
| Slice `arr[i:j]` | O(j-i) | Creates new array |

## Key Insight: Insert/Delete at Front is SLOW

```python
arr = [1, 2, 3, 4, 5]
arr.insert(0, 0)  # O(n) -- shifts ALL elements right!
# [0, 1, 2, 3, 4, 5]

# If you need fast front operations, use collections.deque instead
from collections import deque
d = deque([1, 2, 3, 4, 5])
d.appendleft(0)  # O(1) -- no shifting needed
```

````python
# Array operations and their costs
import time

print("ARRAY OPERATIONS -- Time Complexity Demo")
print("=" * 60)

# O(1) operations
arr = list(range(1_000_000))

start = time.perf_counter()
_ = arr[500_000]  # Access by index
t = time.perf_counter() - start
print(f"Access arr[500000]:     {t:.8f}s  O(1)")

start = time.perf_counter()
arr.append(999)  # Append to end
t = time.perf_counter() - start
print(f"Append to end:          {t:.8f}s  O(1) amortized")

# O(n) operations
start = time.perf_counter()
arr.insert(0, -1)  # Insert at front -- shifts everything!
t = time.perf_counter() - start
print(f"Insert at front:        {t:.6f}s  O(n) -- SLOW!")

start = time.perf_counter()
_ = 999 in arr  # Search
t = time.perf_counter() - start
print(f"Search (x in arr):      {t:.6f}s  O(n)")

# Comparison: deque for front operations
from collections import deque
d = deque(range(1_000_000))

start = time.perf_counter()
d.appendleft(-1)  # O(1) front insert
t = time.perf_counter() - start
print(f"\ndeque.appendleft():     {t:.8f}s  O(1) -- FAST!")
print("\nLesson: Use deque when you need fast front insert/delete")
````

# Section 2: Two Pointers Technique

## What is Two Pointers?

Use two indices (pointers) that move through the array, usually from opposite ends
or at different speeds. Reduces O(n^2) brute force to O(n).

```
Array: [1, 2, 3, 4, 5, 6, 7, 8, 9]
        ^                         ^
       left                     right

Move left right, move right left, until they meet.
```

## When to Use Two Pointers

- Array is **sorted** (or can be sorted)
- Finding a **pair** that satisfies a condition
- **Palindrome** checking
- **Removing duplicates** from sorted array
- **Container with most water** type problems

````python
# Two Pointers -- multiple examples
print("TWO POINTERS TECHNIQUE")
print("=" * 60)

# Example 1: Two Sum in sorted array
def two_sum_sorted(nums, target):
    """Find two numbers in sorted array that sum to target. O(n) time, O(1) space."""
    left, right = 0, len(nums) - 1
    while left < right:
        current_sum = nums[left] + nums[right]
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1   # Need larger sum, move left pointer right
        else:
            right -= 1  # Need smaller sum, move right pointer left
    return []

nums = [1, 2, 3, 4, 6, 8, 9]
print(f"\n1. Two Sum Sorted: {nums}, target=10")
result = two_sum_sorted(nums, 10)
print(f"   Result: indices {result} -> values {nums[result[0]]} + {nums[result[1]]} = 10")

# Example 2: Valid palindrome
def is_palindrome(s):
    """Check if string is palindrome. O(n) time, O(1) space."""
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

print(f"\n2. Palindrome check:")
for word in ["racecar", "hello", "madam", "python"]:
    print(f"   '{word}': {is_palindrome(word)}")

# Example 3: Remove duplicates from sorted array (in-place)
def remove_duplicates(nums):
    """Remove duplicates in-place. Return new length. O(n) time, O(1) space."""
    if not nums:
        return 0
    slow = 0  # Points to last unique element
    for fast in range(1, len(nums)):
        if nums[fast] != nums[slow]:
            slow += 1
            nums[slow] = nums[fast]
    return slow + 1

nums = [1, 1, 2, 2, 3, 4, 4, 5]
new_len = remove_duplicates(nums)
print(f"\n3. Remove duplicates: {nums}")
print(f"   After: {nums[:new_len]} (length={new_len})")
print(f"   In-place! No extra array needed.")
````

# Section 3: Sliding Window Technique

## What is Sliding Window?

Maintain a "window" (subarray/substring) that slides through the data.
Instead of recomputing from scratch each time, update incrementally.

```
Array: [2, 1, 5, 1, 3, 2], window size = 3

Window 1: [2, 1, 5]  sum = 8
Window 2:    [1, 5, 1]  sum = 7  (subtract 2, add 1)
Window 3:       [5, 1, 3]  sum = 9  (subtract 1, add 3)
Window 4:          [1, 3, 2]  sum = 6  (subtract 5, add 2)

Without sliding window: O(n*k) -- recompute each window
With sliding window: O(n) -- just add/subtract one element
```

## Fixed vs Variable Window

- **Fixed size**: Window size is given (e.g., "max sum of k consecutive elements")
- **Variable size**: Window expands/shrinks based on condition (e.g., "longest substring without repeating characters")

````python
# Sliding Window -- fixed and variable size
print("SLIDING WINDOW TECHNIQUE")
print("=" * 60)

# Fixed window: Maximum sum of k consecutive elements
def max_sum_subarray(nums, k):
    """Find max sum of k consecutive elements. O(n) time."""
    if len(nums) < k:
        return 0

    # Compute first window
    window_sum = sum(nums[:k])
    max_sum = window_sum

    # Slide the window
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]  # Add new, remove old
        max_sum = max(max_sum, window_sum)

    return max_sum

nums = [2, 1, 5, 1, 3, 2]
k = 3
print(f"\n1. Max sum of {k} consecutive elements in {nums}:")
print(f"   Result: {max_sum_subarray(nums, k)}")

# Variable window: Longest substring without repeating characters
def length_of_longest_substring(s):
    """Find longest substring without repeating chars. O(n) time."""
    char_index = {}  # {char: last_seen_index}
    left = 0
    max_len = 0

    for right, char in enumerate(s):
        # If char was seen and is within current window, shrink window
        if char in char_index and char_index[char] >= left:
            left = char_index[char] + 1

        char_index[char] = right
        max_len = max(max_len, right - left + 1)

    return max_len

print(f"\n2. Longest substring without repeating characters:")
for s in ["abcabcbb", "bbbbb", "pwwkew", "abcde"]:
    result = length_of_longest_substring(s)
    print(f"   '{s}' -> {result}")

# Variable window: Minimum window substring
def min_window_substring(s, t):
    """Find minimum window in s containing all chars of t. O(n) time."""
    from collections import Counter
    need = Counter(t)
    missing = len(t)
    left = 0
    best = ""

    for right, char in enumerate(s):
        if need[char] > 0:
            missing -= 1
        need[char] -= 1

        if missing == 0:  # Found valid window
            # Shrink from left
            while need[s[left]] < 0:
                need[s[left]] += 1
                left += 1
            window = s[left:right + 1]
            if not best or len(window) < len(best):
                best = window
            need[s[left]] += 1
            missing += 1
            left += 1

    return best

print(f"\n3. Minimum window substring:")
print(f"   min_window('ADOBECODEBANC', 'ABC') = '{min_window_substring('ADOBECODEBANC', 'ABC')}'")
````

# Section 4: Hash Maps -- The Swiss Army Knife

## Why Hash Maps Are So Powerful

A hash map (Python `dict`) gives you O(1) average time for:
- Insert: `d[key] = value`
- Lookup: `d[key]` or `key in d`
- Delete: `del d[key]`

This turns many O(n^2) problems into O(n).

## Common Hash Map Patterns

### Pattern 1: Frequency Counting
```python
# Count how many times each element appears
freq = {}
for x in arr:
    freq[x] = freq.get(x, 0) + 1
# Or: from collections import Counter; freq = Counter(arr)
```

### Pattern 2: Complement Lookup (Two Sum)
```python
# For each element, check if its complement exists
seen = {}
for i, x in enumerate(arr):
    complement = target - x
    if complement in seen:
        return [seen[complement], i]
    seen[x] = i
```

### Pattern 3: Grouping
```python
# Group elements by some key
groups = {}
for item in items:
    key = compute_key(item)
    groups.setdefault(key, []).append(item)
```

````python
# Hash Map patterns -- comprehensive examples
from collections import Counter, defaultdict

print("HASH MAP PATTERNS")
print("=" * 60)

# Pattern 1: Frequency counting
print("\n1. FREQUENCY COUNTING")
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
freq = Counter(words)
print(f"   Words: {words}")
print(f"   Frequencies: {dict(freq)}")
print(f"   Most common: {freq.most_common(2)}")

# Pattern 2: Two Sum (complement lookup)
print("\n2. TWO SUM (complement lookup)")
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

nums = [2, 7, 11, 15]
print(f"   nums={nums}, target=9 -> {two_sum(nums, 9)}")

# Pattern 3: Group Anagrams
print("\n3. GROUP ANAGRAMS")
def group_anagrams(strs):
    groups = defaultdict(list)
    for s in strs:
        key = tuple(sorted(s))  # Sorted chars as key
        groups[key].append(s)
    return list(groups.values())

words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(f"   Input: {words}")
print(f"   Groups: {group_anagrams(words)}")

# Pattern 4: Longest consecutive sequence
print("\n4. LONGEST CONSECUTIVE SEQUENCE")
def longest_consecutive(nums):
    num_set = set(nums)
    max_len = 0
    for num in num_set:
        if num - 1 not in num_set:  # Start of a sequence
            current = num
            length = 1
            while current + 1 in num_set:
                current += 1
                length += 1
            max_len = max(max_len, length)
    return max_len

nums = [100, 4, 200, 1, 3, 2]
print(f"   nums={nums}")
print(f"   Longest consecutive: {longest_consecutive(nums)}")
print(f"   (1, 2, 3, 4 is the longest sequence)")
````

# Section 5: String Manipulation

## Strings in Python

Strings are **immutable** in Python -- you can't modify them in place.
Every string operation creates a new string.

```python
s = "hello"
s[0] = "H"  # ERROR! Strings are immutable

# Instead:
s = "H" + s[1:]  # Creates new string "Hello"
```

## Key String Operations

| Operation | Time | Notes |
|-----------|------|-------|
| `s[i]` | O(1) | Index access |
| `len(s)` | O(1) | Cached |
| `s + t` | O(n+m) | Creates new string |
| `s in t` | O(n*m) | Substring search |
| `s.split()` | O(n) | Creates list |
| `"".join(list)` | O(n) | Efficient concatenation |
| `s[::-1]` | O(n) | Reverse |

## Important: String Concatenation in Loops

```python
# SLOW: O(n^2) -- creates new string each iteration
result = ""
for char in chars:
    result += char  # Each += creates a new string!

# FAST: O(n) -- build list, join once
parts = []
for char in chars:
    parts.append(char)
result = "".join(parts)  # One allocation
```

````python
# String manipulation patterns
print("STRING MANIPULATION PATTERNS")
print("=" * 60)

# Pattern 1: Check if two strings are anagrams
def is_anagram(s, t):
    """O(n) time, O(1) space (26 letters)."""
    if len(s) != len(t):
        return False
    return Counter(s) == Counter(t)

print("\n1. Anagram check:")
pairs = [("anagram", "nagaram"), ("rat", "car"), ("listen", "silent")]
for s, t in pairs:
    print(f"   '{s}' and '{t}': {is_anagram(s, t)}")

# Pattern 2: Reverse words in a string
def reverse_words(s):
    """Reverse order of words. O(n) time."""
    return " ".join(s.split()[::-1])

print("\n2. Reverse words:")
sentences = ["the sky is blue", "  hello world  ", "a good   example"]
for s in sentences:
    print(f"   '{s}' -> '{reverse_words(s)}'")

# Pattern 3: Valid parentheses (stack-based)
def is_valid_parentheses(s):
    """Check if brackets are balanced. O(n) time."""
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}
    for char in s:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()
    return len(stack) == 0

print("\n3. Valid parentheses:")
for s in ["()", "()[]{}", "(]", "([)]", "{[]}"]:
    print(f"   '{s}': {is_valid_parentheses(s)}")

# Pattern 4: Longest palindromic substring
def longest_palindrome(s):
    """Expand around center. O(n^2) time, O(1) space."""
    def expand(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left + 1:right]

    result = ""
    for i in range(len(s)):
        odd = expand(i, i)      # Odd length palindrome
        even = expand(i, i + 1) # Even length palindrome
        if len(odd) > len(result):
            result = odd
        if len(even) > len(result):
            result = even
    return result

print("\n4. Longest palindromic substring:")
for s in ["babad", "cbbd", "racecar", "abcba"]:
    print(f"   '{s}' -> '{longest_palindrome(s)}'")
````

````python
# ============================================
# SOLUTIONS
# ============================================

# Problem 1: Contains Duplicate -- O(n) time, O(n) space
def contains_duplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False
    # One-liner: return len(nums) != len(set(nums))

# Problem 2: Product Except Self -- O(n) time, O(1) extra space
def product_except_self(nums):
    n = len(nums)
    result = [1] * n

    # Left pass: result[i] = product of all elements to the LEFT of i
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]

    # Right pass: multiply by product of all elements to the RIGHT of i
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]

    return result

# Problem 3: Valid Anagram -- O(n) time, O(1) space (26 letters)
def is_anagram(s, t):
    if len(s) != len(t):
        return False
    count = [0] * 26
    for c in s:
        count[ord(c) - ord("a")] += 1
    for c in t:
        count[ord(c) - ord("a")] -= 1
    return all(x == 0 for x in count)

# Test all solutions
print("SOLUTIONS")
print("=" * 60)

print("\n1. Contains Duplicate:")
tests = [([1,2,3,1], True), ([1,2,3,4], False), ([1,1,1,3,3,4,3,2,4,2], True)]
for nums, expected in tests:
    result = contains_duplicate(nums)
    status = "PASS" if result == expected else "FAIL"
    print(f"   [{status}] {nums} -> {result}")

print("\n2. Product Except Self:")
tests = [([1,2,3,4], [24,12,8,6]), ([2,3,4,5], [60,40,30,24])]
for nums, expected in tests:
    result = product_except_self(nums)
    status = "PASS" if result == expected else "FAIL"
    print(f"   [{status}] {nums} -> {result}")

print("\n3. Valid Anagram:")
tests = [("anagram","nagaram",True), ("rat","car",False), ("listen","silent",True)]
for s, t, expected in tests:
    result = is_anagram(s, t)
    status = "PASS" if result == expected else "FAIL"
    print(f"   [{status}] '{s}','{t}' -> {result}")
````
