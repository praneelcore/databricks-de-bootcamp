# 📗 PART 1: What is Data Engineering?

## Section 1.1: Definition & Evolution

### What is Data Engineering?

**Data Engineering** is the discipline of designing, building, and maintaining the systems
and infrastructure that collect, store, transform, and serve data at scale.

Think of it this way:
- **Data Scientists** are the chefs who create amazing dishes (insights, models)
- **Data Engineers** are the ones who build the kitchen, stock the pantry, maintain the
  equipment, and ensure fresh ingredients arrive on time every day

Without data engineering, data science is impossible. You can't train a model on data
that doesn't exist in a usable form.

### The Evolution of Data Engineering

```
1970s-1990s: Database Administrators (DBAs)
  └── Managed single relational databases (Oracle, DB2)
  └── Focus: uptime, backups, query tuning

2000s: ETL Developers
  └── Built Extract-Transform-Load pipelines
  └── Tools: Informatica, SSIS, DataStage
  └── Focus: moving data between systems

2010s: Big Data Engineers
  └── Hadoop, MapReduce, Hive, Pig
  └── Focus: processing massive datasets (petabytes)

2015-2020: Modern Data Engineers
  └── Cloud-native (AWS, Azure, GCP)
  └── Spark, Kafka, Airflow
  └── Focus: real-time + batch, cloud scale

2020-Present: Platform Data Engineers
  └── Lakehouses, streaming-first, DataOps
  └── Databricks, Snowflake, dbt
  └── Focus: self-serve platforms, data products, governance
```

## Section 2.2: Ingestion (Getting Data In)

Ingestion is the process of moving data from source systems into your data platform.
This is often the most fragile part of a pipeline — source systems change without warning.

### Batch vs Streaming Ingestion

| Aspect | Batch Ingestion | Streaming Ingestion |
|--------|----------------|-------------------|
| **Frequency** | Hourly, daily, weekly | Continuous (seconds) |
| **Latency** | Minutes to hours | Milliseconds to seconds |
| **Complexity** | Lower | Higher |
| **Cost** | Lower (scheduled compute) | Higher (always-on compute) |
| **Use cases** | Reports, analytics, ML training | Real-time dashboards, alerts, fraud |
| **Tools** | Airflow, dbt, Spark batch | Kafka, Kinesis, Flink, Spark Streaming |
| **Error handling** | Retry entire batch | Dead letter queues, checkpoints |

### Ingestion Patterns

**1. Full Load (Replace)**
```
Source Table (1M rows) ──── COPY ALL ────▶ Target Table (replace)
```
- Simple but expensive at scale
- Use when: small tables, no reliable change tracking

**2. Incremental Load (Append New)**
```
Source Table ──── WHERE modified_date > last_run ────▶ Target Table (append)
```
- Efficient but requires a reliable watermark column
- Use when: append-only data (logs, events)

**3. Change Data Capture (CDC)**
```
Source DB ──── Transaction Log ────▶ CDC Stream ────▶ Target (MERGE)
```
- Captures inserts, updates, AND deletes
- Use when: you need the full picture including deletions

**4. Event Streaming**
```
Producer ──── Kafka/Kinesis ────▶ Consumer ────▶ Target
```
- Real-time, decoupled architecture
- Use when: multiple consumers need the same data

### Production Example: Airbnb's Data Ingestion

Airbnb ingests data from 100+ source systems:
- **MySQL databases** (listings, bookings, users) → CDC via Debezium → Kafka → S3
- **Application logs** (search, pricing, messaging) → Fluentd → Kafka → S3
- **Third-party APIs** (Stripe payments, Twilio SMS) → Custom Python jobs → S3
- **Event streams** (user behavior, experiments) → Kafka → S3

All data lands in S3 as Parquet files, partitioned by date.

````python
# Demonstrating ingestion patterns
from datetime import datetime, timedelta

class DataIngestionSimulator:
    """Simulates different ingestion patterns."""
    
    def __init__(self):
        self.target_data = []
        self.last_run_timestamp = None
    
    def full_load(self, source_data):
        """Pattern 1: Full load — replace everything."""
        print("🔄 FULL LOAD Pattern")
        print(f"   Source records: {len(source_data)}")
        
        # Replace entire target
        self.target_data = source_data.copy()
        
        print(f"   Target records after load: {len(self.target_data)}")
        print(f"   Strategy: DROP + INSERT (or OVERWRITE)")
        print(f"   ⚠️ Cost: Copies ALL data every time")
        print(f"   ✅ Simplicity: No change tracking needed")
        return self.target_data
    
    def incremental_load(self, source_data, watermark_column="updated_at"):
        """Pattern 2: Incremental — only new/changed records."""
        print("\n📥 INCREMENTAL LOAD Pattern")
        
        if self.last_run_timestamp is None:
            # First run = full load
            self.target_data = source_data.copy()
            self.last_run_timestamp = max(r[watermark_column] for r in source_data)
            print(f"   First run (full load): {len(source_data)} records")
        else:
            # Subsequent runs = only new records
            new_records = [r for r in source_data 
                         if r[watermark_column] > self.last_run_timestamp]
            self.target_data.extend(new_records)
            self.last_run_timestamp = max(r[watermark_column] for r in source_data)
            print(f"   Incremental: {len(new_records)} new records (of {len(source_data)} total)")
        
        print(f"   Target total: {len(self.target_data)} records")
        print(f"   Watermark: {watermark_column} > {self.last_run_timestamp}")
        return self.target_data
    
    def cdc_load(self, cdc_events):
        """Pattern 3: CDC — apply inserts, updates, and deletes."""
        print("\n🔀 CDC (Change Data Capture) Pattern")
        
        inserts = updates = deletes = 0
        target_dict = {r["id"]: r for r in self.target_data}
        
        for event in cdc_events:
            op = event["operation"]
            if op == "INSERT":
                target_dict[event["data"]["id"]] = event["data"]
                inserts += 1
            elif op == "UPDATE":
                if event["data"]["id"] in target_dict:
                    target_dict[event["data"]["id"]] = event["data"]
                    updates += 1
            elif op == "DELETE":
                if event["data"]["id"] in target_dict:
                    del target_dict[event["data"]["id"]]
                    deletes += 1
        
        self.target_data = list(target_dict.values())
        print(f"   CDC events processed: {len(cdc_events)}")
        print(f"   Inserts: {inserts}, Updates: {updates}, Deletes: {deletes}")
        print(f"   Target records: {len(self.target_data)}")
        return self.target_data


# Demo all three patterns
simulator = DataIngestionSimulator()

# Source data
source = [
    {"id": 1, "name": "Alice", "amount": 100, "updated_at": "2024-03-01"},
    {"id": 2, "name": "Bob", "amount": 200, "updated_at": "2024-03-01"},
    {"id": 3, "name": "Charlie", "amount": 300, "updated_at": "2024-03-02"},
]

# Pattern 1: Full Load
print("=" * 60)
simulator.full_load(source)

# Pattern 2: Incremental
print("\n" + "=" * 60)
sim2 = DataIngestionSimulator()
sim2.incremental_load(source)
# Simulate new data arriving
source_updated = source + [
    {"id": 4, "name": "Diana", "amount": 400, "updated_at": "2024-03-03"},
]
sim2.incremental_load(source_updated)

# Pattern 3: CDC
print("\n" + "=" * 60)
sim3 = DataIngestionSimulator()
sim3.target_data = source.copy()  # Start with existing data
cdc_events = [
    {"operation": "INSERT", "data": {"id": 4, "name": "Diana", "amount": 400}},
    {"operation": "UPDATE", "data": {"id": 1, "name": "Alice", "amount": 150}},
    {"operation": "DELETE", "data": {"id": 2, "name": "Bob", "amount": 200}},
]
sim3.cdc_load(cdc_events)
````

## Section 2.3: Storage (Where Data Lives)

Storage is where data rests between processing stages. The choice of storage system
fundamentally shapes your architecture's capabilities and limitations.

### The Storage Spectrum

```
         Raw / Flexible                              Structured / Optimized
         ◄──────────────────────────────────────────────────────────────────►
         
┌──────────────┐  ┌──────────────┐  ┌──────────────┐  ┌──────────────┐
│  Object Store │  │  Data Lake   │  │  Lakehouse   │  │  Warehouse   │
│  (S3, GCS)   │  │  (Lake +     │  │  (Lake +     │  │  (Snowflake, │
│              │  │   Catalog)   │  │   ACID +     │  │   BigQuery)  │
│  Any format  │  │  Parquet/ORC │  │   Schema)    │  │  Columnar    │
│  No schema   │  │  Schema-on-  │  │  Schema +    │  │  Schema      │
│  Cheapest    │  │  read        │  │  ACID        │  │  enforced    │
│              │  │              │  │  (Delta,     │  │  Most        │

Raw data is rarely useful as-is. Transformation is where data engineers add the most value —

### ETL vs ELT

```
ETL (Traditional):                        ELT (Modern):
┌────────┐  ┌───────────┐  ┌──────┐     ┌────────┐  ┌──────┐  ┌───────────┐
│Extract │─▶│ Transform │─▶│ Load │     │Extract │─▶│ Load │─▶│ Transform │
│        │  │(external) │  │      │     │        │  │      │  │(in-place) │

```

### The Medallion Architecture (Bronze → Silver → Gold)

The most popular transformation pattern in modern data engineering:

```
┌─────────────────┐    ┌─────────────────┐    ┌─────────────────┐
│     BRONZE      │    │     SILVER      │    │      GOLD       │
│   (Raw Data)    │───▶│ (Cleaned Data)  │───▶│ (Business Data) │
│                 │    │                 │    │                 │
│ • Exact copy    │    │ • Deduplicated  │    │ • Aggregated    │
│ • All formats   │    │ • Type-cast     │    │ • Joined        │
│ • Append-only   │    │ • Validated     │    │ • Business logic│
│ • Schema-on-    │    │ • Standardized  │    │ • Ready for BI  │
│   read          │    │ • Filtered      │    │ • Ready for ML  │
└─────────────────┘    └─────────────────┘    └─────────────────┘
```

### Transformation Types

| Type | What It Does | Example |
|------|-------------|---------|
| **Cleaning** | Fix data quality issues | Remove nulls, trim whitespace, fix dates |
| **Deduplication** | Remove duplicate records | Keep latest record per customer_id |
| **Type Casting** | Convert data types | String "2024-01-15" → DATE type |
| **Enrichment** | Add context from other sources | Join orders with customer segments |
| **Aggregation** | Summarize data | Daily revenue = SUM(order_total) GROUP BY date |
| **Normalization** | Standardize values | "USA", "US", "United States" → "US" |
| **Denormalization** | Flatten for performance | Join fact + dimensions into wide table |
| **Derivation** | Calculate new fields | profit = revenue - cost |
| **Filtering** | Remove irrelevant data | WHERE status != 'test' |
| **Pivoting** | Reshape data structure | Rows → Columns or Columns → Rows |

```
Raw Listens (Bronze)          User Preferences (Silver)       Discover Weekly (Gold)

├── track_id                  ├── genre_preferences[]         ├── playlist_name

Billions of raw events → ML-ready features → Personalized playlists

````python
# Demonstrating the Medallion Architecture transformation pattern
import json

# ===== BRONZE LAYER: Raw data (as-is from source) =====
bronze_orders = [
    {"order_id": "1001", "customer_id": "42", "amount": "89.99", 
     "status": "COMPLETED", "date": "2024-03-15", "source": "web"},
    {"order_id": "1001", "customer_id": "42", "amount": "89.99",  # DUPLICATE!
     "status": "COMPLETED", "date": "2024-03-15", "source": "web"},
    {"order_id": "1002", "customer_id": "17", "amount": "invalid",  # BAD DATA!
     "status": "pending", "date": "03/15/2024", "source": "mobile"},
    {"order_id": "1003", "customer_id": None, "amount": "300.00",  # NULL customer!
     "status": "completed", "date": "2024-03-15", "source": "api"},
    {"order_id": "1004", "customer_id": "99", "amount": "45.50",
     "status": "Completed", "date": "2024-03-16", "source": "web"},
]

print("🥉 BRONZE LAYER (Raw — warts and all)")
print(f"   Records: {len(bronze_orders)}")
print(f"   Issues: duplicates, nulls, inconsistent formats, bad data")
for r in bronze_orders[:3]:
    print(f"   {r}")

# ===== SILVER LAYER: Cleaned, deduplicated, validated =====
def bronze_to_silver(bronze_records):
    """Transform bronze → silver: clean, deduplicate, validate."""
    seen_ids = set()
    silver_records = []
    rejected = []
    
    for record in bronze_records:
        # 1. Deduplicate
        if record["order_id"] in seen_ids:
            rejected.append({"record": record, "reason": "duplicate"})
            continue
        seen_ids.add(record["order_id"])
        
        # 2. Validate required fields
        if record["customer_id"] is None:
            rejected.append({"record": record, "reason": "null customer_id"})
            continue
        
        # 3. Type casting and validation
        try:
            amount = float(record["amount"])
        except (ValueError, TypeError):
            rejected.append({"record": record, "reason": "invalid amount"})
            continue
        
        # 4. Standardize values
        silver_record = {
            "order_id": int(record["order_id"]),
            "customer_id": int(record["customer_id"]),
            "amount": amount,
            "status": record["status"].lower().strip(),
            "order_date": record["date"],  # Would normalize date format
            "source": record["source"].lower(),
        }
        silver_records.append(silver_record)
    
    return silver_records, rejected

silver_orders, rejected = bronze_to_silver(bronze_orders)

print(f"\n🥈 SILVER LAYER (Cleaned & Validated)")
print(f"   Records: {len(silver_orders)} (from {len(bronze_orders)} bronze)")
print(f"   Rejected: {len(rejected)}")
for r in silver_orders:
    print(f"   {r}")
print(f"   Rejection reasons:")
for r in rejected:
    print(f"   ❌ {r['reason']}: order_id={r['record']['order_id']}")

# ===== GOLD LAYER: Business-ready aggregates =====
def silver_to_gold(silver_records):
    """Transform silver → gold: aggregate for business use."""
    daily_revenue = {}
    for record in silver_records:
        date = record["order_date"]
        if date not in daily_revenue:
            daily_revenue[date] = {"date": date, "orders": 0, "revenue": 0.0}
        daily_revenue[date]["orders"] += 1
        daily_revenue[date]["revenue"] += record["amount"]
    
    # Add derived metrics
    for date, metrics in daily_revenue.items():
        metrics["avg_order_value"] = round(metrics["revenue"] / metrics["orders"], 2)
    
    return list(daily_revenue.values())

gold_daily = silver_to_gold(silver_orders)

print(f"\n🥇 GOLD LAYER (Business-Ready)")
print(f"   Aggregated records: {len(gold_daily)}")
for r in gold_daily:
    print(f"   📊 {r['date']}: {r['orders']} orders, ${r['revenue']:.2f} revenue, "
          f"${r['avg_order_value']:.2f} avg")

print("\n💡 Key Pattern: Bronze (raw) → Silver (clean) → Gold (business)")
print("   Each layer adds value and reduces volume.")
````

## Section 2.6: Orchestration (Coordinating Pipelines)

In production, you don't run pipelines manually. Orchestration tools coordinate
when pipelines run, in what order, and what happens when they fail.

### What is Orchestration?

Think of it like a conductor leading an orchestra:
- Each musician (pipeline) knows their part
- The conductor (orchestrator) ensures they play in the right order
- If a musician makes a mistake, the conductor decides what to do

### DAGs (Directed Acyclic Graphs)

Every orchestration system uses DAGs to represent pipeline dependencies:

```
        ┌─────────────┐
        │ ingest_orders│
        └──────┬──────┘
               │
        ┌──────▼──────┐
        │ clean_orders │
        └──────┬──────┘
               │
    ┌──────────┼──────────┐
    │          │          │
┌───▼───┐ ┌───▼───┐ ┌───▼───┐
│ daily │ │ weekly│ │  ML   │
│revenue│ │cohort │ │features│
└───┬───┘ └───┬───┘ └───┬───┘
    │          │          │
    └──────────┼──────────┘
               │
        ┌──────▼──────┐
        │  send_report │
        └─────────────┘
```

Rules of a DAG:
- **Directed**: Edges have direction (A → B means A runs before B)
- **Acyclic**: No circular dependencies (A → B → C → A is INVALID)
- **Graph**: Nodes (tasks) connected by edges (dependencies)

### Orchestration Tools Comparison

| Tool | Philosophy | Best For | Learning Curve |
|------|-----------|----------|---------------|
| **Apache Airflow** | Code-first (Python DAGs) | Complex workflows, large teams | Medium |
| **Prefect** | Pythonic, modern | Python-heavy teams | Low |
| **Dagster** | Asset-oriented | Data-aware orchestration | Medium |
| **Databricks Workflows** | Integrated with Databricks | Databricks-native teams | Low |
| **AWS Step Functions** | Serverless, JSON-based | AWS-native, simple flows | Low |
| **dbt Cloud** | SQL transformation scheduling | dbt-only workflows | Low |

### Production Example: Airbnb's Orchestration

Airbnb runs 10,000+ Airflow DAGs daily:
- **Search ranking pipeline**: Runs every 4 hours, 200+ tasks
- **Financial reconciliation**: Runs nightly, must complete by 6 AM
- **Experiment analysis**: Triggered by experiment completion events
- **ML model retraining**: Weekly, depends on feature pipeline completion

````python
# Simulating a DAG-based orchestration system
from datetime import datetime
import time

class Task:
    """Represents a single task in a DAG."""
    def __init__(self, name, dependencies=None):
        self.name = name
        self.dependencies = dependencies or []
        self.status = "pending"  # pending, running, success, failed
        self.start_time = None
        self.end_time = None
    
    def run(self):
        self.status = "running"
        self.start_time = datetime.now()
        # Simulate work
        self.status = "success"
        self.end_time = datetime.now()
        return True

class DAG:
    """Simple DAG orchestrator."""
    def __init__(self, name, schedule="daily"):
        self.name = name
        self.schedule = schedule
        self.tasks = {}
    
    def add_task(self, task):
        self.tasks[task.name] = task
    
    def get_execution_order(self):
        """Topological sort — determine correct execution order."""
        visited = set()
        order = []
        
        def visit(task_name):
            if task_name in visited:
                return
            visited.add(task_name)
            task = self.tasks[task_name]
            for dep in task.dependencies:
                visit(dep)
            order.append(task_name)
        
        for task_name in self.tasks:
            visit(task_name)
        return order
    
    def execute(self):
        """Execute all tasks in dependency order."""
        order = self.get_execution_order()
        print(f"🎼 Executing DAG: {self.name}")
        print(f"   Schedule: {self.schedule}")
        print(f"   Tasks: {len(self.tasks)}")
        print(f"   Execution order: {' → '.join(order)}")
        print()
        
        for task_name in order:
            task = self.tasks[task_name]
            deps_status = [self.tasks[d].status for d in task.dependencies]
            
            if all(s == "success" for s in deps_status):
                task.run()
                print(f"   ✅ {task_name} — completed")
            else:
                task.status = "skipped"
                print(f"   ⏭️ {task_name} — skipped (dependency failed)")


# Build a sample ETL DAG
etl_dag = DAG("daily_revenue_pipeline", schedule="0 6 * * *")  # 6 AM daily

# Define tasks with dependencies
etl_dag.add_task(Task("ingest_orders"))
etl_dag.add_task(Task("ingest_customers"))
etl_dag.add_task(Task("ingest_products"))
etl_dag.add_task(Task("clean_orders", dependencies=["ingest_orders"]))
etl_dag.add_task(Task("clean_customers", dependencies=["ingest_customers"]))
etl_dag.add_task(Task("join_order_details", dependencies=["clean_orders", "ingest_products"]))
etl_dag.add_task(Task("build_daily_revenue", dependencies=["join_order_details"]))
etl_dag.add_task(Task("build_customer_segments", dependencies=["clean_customers", "join_order_details"]))
etl_dag.add_task(Task("send_slack_report", dependencies=["build_daily_revenue", "build_customer_segments"]))

# Execute
etl_dag.execute()

print("\n💡 Key Concepts:")
print("   • Tasks run in dependency order (topological sort)")
print("   • Independent tasks CAN run in parallel (ingest_orders || ingest_customers)")
print("   • If a task fails, downstream tasks are skipped")
print("   • The DAG is idempotent — safe to rerun")
````

## Section 2.7: Monitoring & Observability

You can't fix what you can't see. Monitoring is what separates hobby projects from production systems.

### The Three Pillars of Data Observability

| Pillar | What It Tracks | Tools |
|--------|---------------|-------|
| **Data Quality** | Is the data correct? (completeness, accuracy, freshness) | Great Expectations, Monte Carlo, Soda |
| **Pipeline Health** | Are pipelines running? (success/failure, duration, cost) | Airflow UI, Datadog, CloudWatch |
| **Data Lineage** | Where did data come from? Where does it go? | OpenLineage, Marquez, Unity Catalog |

### Key Metrics to Monitor

**Pipeline Metrics:**
- ✅ Success/failure rate (target: >99.5%)
- ⏱️ Execution duration (detect slowdowns)
- 💰 Cost per run (prevent budget surprises)
- 📊 Records processed (detect volume anomalies)

**Data Quality Metrics:**
- 📋 Completeness: % of non-null values (target: >99.9%)
- 🎯 Accuracy: % of values passing validation rules
- ⏰ Freshness: Time since last update (SLA compliance)
- 🔢 Volume: Row count within expected range
- 🔑 Uniqueness: No unexpected duplicates

### Production Example: Netflix's Data Quality Framework

Netflix monitors 10,000+ datasets with automated checks:
- **Freshness alerts**: If a dataset isn't updated within its SLA, page the owner
- **Volume anomalies**: If row count drops >20% from yesterday, alert
- **Schema changes**: If a column is added/removed, notify downstream consumers
- **Value distribution**: If NULL rate spikes from 0.1% to 5%, investigate

````python
# Data quality monitoring framework
from datetime import datetime, timedelta
import random

class DataQualityMonitor:
    """Production-style data quality monitoring."""
    
    def __init__(self, table_name):
        self.table_name = table_name
        self.checks = []
        self.results = []
    
    def add_check(self, name, check_type, params):
        self.checks.append({"name": name, "type": check_type, "params": params})
    
    def run_checks(self, data):
        """Run all quality checks against the data."""
        print(f"🔍 Running quality checks on: {self.table_name}")
        print(f"   Records: {len(data):,}")
        print(f"   Checks: {len(self.checks)}")
        print("-" * 50)
        
        passed = 0
        failed = 0
        
        for check in self.checks:
            result = self._execute_check(check, data)
            self.results.append(result)
            
            status = "✅ PASS" if result["passed"] else "❌ FAIL"
            print(f"   {status} | {check['name']}: {result['message']}")
            
            if result["passed"]:
                passed += 1
            else:
                failed += 1
        
        print("-" * 50)
        print(f"   Results: {passed} passed, {failed} failed")
        print(f"   Score: {passed/(passed+failed)*100:.1f}%")
        
        if failed > 0:
            print("   🚨 ACTION REQUIRED: Fix failing checks before serving data!")
        else:
            print("   ✅ All checks passed — data is ready for consumption")
        
        return failed == 0
    
    def _execute_check(self, check, data):
        """Execute a single quality check."""
        check_type = check["type"]
        params = check["params"]
        
        if check_type == "not_null":
            column = params["column"]
            null_count = sum(1 for r in data if r.get(column) is None)
            null_pct = null_count / len(data) * 100
            threshold = params.get("threshold_pct", 0)
            passed = null_pct <= threshold
            return {"passed": passed, "message": f"{column}: {null_pct:.2f}% nulls (threshold: {threshold}%)"}
        
        elif check_type == "unique":
            column = params["column"]
            values = [r[column] for r in data if r.get(column) is not None]
            unique_count = len(set(values))
            dup_count = len(values) - unique_count
            passed = dup_count == 0
            return {"passed": passed, "message": f"{column}: {dup_count} duplicates found"}
        
        elif check_type == "range":
            column = params["column"]
            min_val = params.get("min")
            max_val = params.get("max")
            values = [r[column] for r in data if r.get(column) is not None]
            out_of_range = sum(1 for v in values if v < min_val or v > max_val)
            passed = out_of_range == 0
            return {"passed": passed, "message": f"{column}: {out_of_range} values outside [{min_val}, {max_val}]"}
        
        elif check_type == "row_count":
            min_rows = params.get("min", 0)
            max_rows = params.get("max", float("inf"))
            count = len(data)
            passed = min_rows <= count <= max_rows
            return {"passed": passed, "message": f"Row count: {count:,} (expected: {min_rows:,}-{max_rows:,})"}
        
        return {"passed": True, "message": "Unknown check type"}


# Demo: Monitor a dataset
sample_data = [
    {"order_id": i, "customer_id": i % 100 if i % 50 != 0 else None,
     "amount": random.uniform(5, 500) if i % 30 != 0 else -10,
     "status": random.choice(["completed", "pending", "failed"])}
    for i in range(1, 1001)
]
# Add a duplicate
sample_data.append({"order_id": 1, "customer_id": 1, "amount": 50, "status": "completed"})

monitor = DataQualityMonitor("gold.daily_orders")
monitor.add_check("Order ID uniqueness", "unique", {"column": "order_id"})
monitor.add_check("Customer ID not null", "not_null", {"column": "customer_id", "threshold_pct": 1.0})
monitor.add_check("Amount positive", "range", {"column": "amount", "min": 0, "max": 10000})
monitor.add_check("Row count in range", "row_count", {"min": 900, "max": 1100})

monitor.run_checks(sample_data)
````

## Section 3.2: Data Warehousing

A data warehouse is a centralized repository optimized for analytical queries.
It's where business users go to answer questions about the business.

### OLTP vs OLAP

| Aspect | OLTP (Transactional) | OLAP (Analytical) |
|--------|---------------------|-------------------|
| **Purpose** | Run the business | Analyze the business |
| **Users** | Applications, customers | Analysts, data scientists |
| **Queries** | Simple (get order #123) | Complex (revenue by region by quarter) |
| **Data** | Current state only | Historical (years of data) |
| **Volume** | Thousands of rows/query | Millions to billions/query |
| **Optimization** | Fast writes (INSERT/UPDATE) | Fast reads (SELECT/GROUP BY) |
| **Storage** | Row-oriented | Column-oriented |
| **Examples** | MySQL, PostgreSQL | Snowflake, BigQuery, Redshift |

### Slowly Changing Dimensions (SCD)

Dimensions change over time. How do you track history?

| SCD Type | Strategy | Example | Trade-off |
|----------|----------|---------|-----------|
| **Type 0** | Never update | Birth date | Immutable facts only |
| **Type 1** | Overwrite | Fix typo in name | Lose history |
| **Type 2** | Add new row + versioning | Customer moves cities | Full history, complex |
| **Type 3** | Add column for previous | Previous + current address | Limited history |
| **Type 6** | Hybrid (1+2+3) | Current flag + history + previous | Most flexible, most complex |

### Production Example: Walmart's Data Warehouse

Walmart operates one of the world's largest data warehouses:
- **Scale**: 40+ petabytes of structured data
- **Users**: 10,000+ analysts and data scientists
- **Queries**: 1M+ queries per day
- **Freshness**: Near real-time for inventory, daily for financials
- **Use cases**: Inventory optimization, pricing, supply chain, customer analytics

````python
# SCD Type 2 implementation — tracking dimension history
from datetime import date

class SCDType2Dimension:
    """Implements Slowly Changing Dimension Type 2."""
    
    def __init__(self, name):
        self.name = name
        self.records = []
        self.surrogate_key = 0
    
    def _next_key(self):
        self.surrogate_key += 1
        return self.surrogate_key
    
    def insert(self, business_key, attributes, effective_date):
        """Insert a new dimension record."""
        record = {
            "surrogate_key": self._next_key(),
            "business_key": business_key,
            **attributes,
            "effective_from": effective_date,
            "effective_to": date(9999, 12, 31),  # "infinity"
            "is_current": True
        }
        self.records.append(record)
        return record
    
    def update(self, business_key, new_attributes, change_date):
        """SCD Type 2: Close current record, insert new version."""
        # Find current record
        current = None
        for r in self.records:
            if r["business_key"] == business_key and r["is_current"]:
                current = r
                break
        
        if current is None:
            return self.insert(business_key, new_attributes, change_date)
        
        # Close current record
        current["effective_to"] = change_date
        current["is_current"] = False
        
        # Insert new version
        new_record = {
            "surrogate_key": self._next_key(),
            "business_key": business_key,
            **new_attributes,
            "effective_from": change_date,
            "effective_to": date(9999, 12, 31),
            "is_current": True
        }
        self.records.append(new_record)
        return new_record
    
    def get_current(self, business_key):
        """Get the current version of a dimension record."""
        for r in self.records:
            if r["business_key"] == business_key and r["is_current"]:
                return r
        return None
    
    def get_history(self, business_key):
        """Get all versions of a dimension record."""
        return [r for r in self.records if r["business_key"] == business_key]


# Demo: Track a customer who moves cities
customer_dim = SCDType2Dimension("dim_customers")

# Initial load
customer_dim.insert("CUST-001", {"name": "Alice Smith", "city": "New York", "segment": "Standard"}, date(2022, 1, 1))
customer_dim.insert("CUST-002", {"name": "Bob Jones", "city": "Los Angeles", "segment": "Premium"}, date(2022, 1, 1))

# Alice moves to San Francisco (SCD Type 2 change)
customer_dim.update("CUST-001", {"name": "Alice Smith", "city": "San Francisco", "segment": "Standard"}, date(2023, 6, 15))

# Alice gets upgraded to Premium
customer_dim.update("CUST-001", {"name": "Alice Smith", "city": "San Francisco", "segment": "Premium"}, date(2024, 1, 1))

print("📊 SCD TYPE 2 — Customer Dimension History")
print("=" * 80)
print(f"{'SK':<4} {'Business Key':<12} {'Name':<15} {'City':<15} {'Segment':<10} {'From':<12} {'To':<12} {'Current'}")
print("-" * 80)
for r in customer_dim.records:
    print(f"{r['surrogate_key']:<4} {r['business_key']:<12} {r['name']:<15} "
          f"{r['city']:<15} {r['segment']:<10} {str(r['effective_from']):<12} "
          f"{str(r['effective_to']):<12} {'✅' if r['is_current'] else '❌'}")

print("\n🔍 Alice's History:")
for r in customer_dim.get_history("CUST-001"):
    print(f"   {r['effective_from']} → {r['effective_to']}: {r['city']}, {r['segment']}")

print("\n💡 SCD Type 2 preserves FULL history — you can query 'as of' any date")
print("   Query: 'What was Alice's city on 2023-01-01?' → New York")
print("   Query: 'What is Alice's current city?' → San Francisco")
````

## Section 3.3: Data Lakes & Lakehouses

### Data Lake

A data lake stores ALL data in its raw format — structured, semi-structured, and unstructured.
Think of it as a massive, cheap storage system where you dump everything first, then process later.

**Advantages:**
- Store anything (CSV, JSON, images, videos, logs)
- Extremely cheap (object storage: $0.023/GB/month on S3)
- Schema-on-read (decide structure when you query, not when you store)
- Scales infinitely

**Disadvantages (the "Data Swamp" problem):**
- No ACID transactions (concurrent writes can corrupt data)
- No schema enforcement (garbage in, garbage stays)
- No indexing (full scans for every query)
- Hard to find data (no catalog)
- Stale data accumulates

### Lakehouse = Data Lake + Warehouse Features

The lakehouse solves the data swamp problem by adding warehouse features to the lake:

```
┌─────────────────────────────────────────────────────────────┐
│                      LAKEHOUSE                               │
│                                                              │
│  ┌─────────────────────────────────────────────────────┐    │
│  │              Table Format Layer                       │    │
│  │         (Delta Lake / Iceberg / Hudi)                │    │
│  │                                                      │    │
│  │  ✅ ACID Transactions    ✅ Schema Enforcement       │    │
│  │  ✅ Time Travel          ✅ Indexing & Stats         │    │
│  │  ✅ Upserts (MERGE)      ✅ Partition Evolution      │    │
│  │  ✅ Audit History         ✅ Concurrent Access        │    │
│  └─────────────────────────────────────────────────────┘    │
│                           │                                  │
│  ┌─────────────────────────────────────────────────────┐    │
│  │              Object Storage Layer                     │    │
│  │            (S3 / ADLS / GCS)                         │    │
│  │                                                      │    │
│  │  💰 Cheap storage ($0.023/GB/month)                  │    │
│  │  📈 Infinite scale                                   │    │
│  │  📁 Open formats (Parquet files)                     │    │
│  └─────────────────────────────────────────────────────┘    │
│                                                              │
└─────────────────────────────────────────────────────────────┘
```

### Delta Lake vs Iceberg vs Hudi

| Feature | Delta Lake | Apache Iceberg | Apache Hudi |
|---------|-----------|---------------|-------------|
| **Creator** | Databricks | Netflix → Apache | Uber → Apache |
| **Primary use** | Lakehouse platform | Multi-engine analytics | Incremental processing |
| **ACID** | ✅ | ✅ | ✅ |
| **Time Travel** | ✅ | ✅ | ✅ |
| **Schema Evolution** | ✅ | ✅ (excellent) | ✅ |
| **Partition Evolution** | Limited | ✅ (hidden partitioning) | ✅ |
| **Engine Support** | Spark, Databricks | Spark, Trino, Flink, Dremio | Spark, Flink, Presto |
| **Ecosystem** | Databricks-centric | Multi-engine | Streaming-focused |

### Production Example: LinkedIn's Data Lake

LinkedIn manages 100+ petabytes across their data lake:
- **Storage**: Azure Data Lake Storage (ADLS)
- **Format**: Migrating from Avro → Iceberg
- **Catalog**: Custom metadata service
- **Access**: 10,000+ datasets, 5,000+ daily users
- **Governance**: Automated PII detection and masking

````python
# Lakehouse features demonstration
class LakehouseTable:
    """Simulates Delta Lake / Iceberg table features."""
    
    def __init__(self, name, schema):
        self.name = name
        self.schema = schema
        self.data = []
        self.versions = []  # Time travel history
        self.current_version = 0
        self.transaction_log = []
    
    def insert(self, records):
        """ACID insert with versioning."""
        # Validate schema
        for record in records:
            for col in self.schema:
                if col not in record:
                    raise ValueError(f"Missing column: {col}")
        
        self.data.extend(records)
        self.current_version += 1
        self.versions.append({"version": self.current_version, "data": self.data.copy(),
                             "operation": "INSERT", "rows_affected": len(records)})
        self.transaction_log.append(f"v{self.current_version}: INSERT {len(records)} rows")
        return len(records)
    
    def merge(self, source_records, key_column):
        """MERGE (upsert) — insert new, update existing."""
        inserts = 0
        updates = 0
        existing_keys = {r[key_column]: i for i, r in enumerate(self.data)}
        
        for record in source_records:
            key = record[key_column]
            if key in existing_keys:
                self.data[existing_keys[key]] = record
                updates += 1
            else:
                self.data.append(record)
                inserts += 1
        
        self.current_version += 1
        self.versions.append({"version": self.current_version, "data": self.data.copy(),
                             "operation": "MERGE", "rows_affected": inserts + updates})
        self.transaction_log.append(f"v{self.current_version}: MERGE {inserts} inserts, {updates} updates")
        return inserts, updates
    
    def time_travel(self, version):
        """Query data as of a specific version."""
        for v in self.versions:
            if v["version"] == version:
                return v["data"]
        raise ValueError(f"Version {version} not found")
    
    def history(self):
        """Show table history (like DESCRIBE HISTORY in Delta)."""
        return self.transaction_log


# Demo lakehouse features
print("🏠 LAKEHOUSE TABLE FEATURES DEMO")
print("=" * 60)

# Create table
orders_table = LakehouseTable("orders", ["order_id", "customer", "amount", "status"])

# Version 1: Initial load
orders_table.insert([
    {"order_id": 1, "customer": "Alice", "amount": 100, "status": "pending"},
    {"order_id": 2, "customer": "Bob", "amount": 200, "status": "pending"},
    {"order_id": 3, "customer": "Charlie", "amount": 300, "status": "pending"},
])
print("v1: Initial load (3 orders)")

# Version 2: MERGE — update existing + insert new
orders_table.merge([
    {"order_id": 1, "customer": "Alice", "amount": 100, "status": "completed"},  # UPDATE
    {"order_id": 2, "customer": "Bob", "amount": 200, "status": "shipped"},      # UPDATE
    {"order_id": 4, "customer": "Diana", "amount": 400, "status": "pending"},    # INSERT
], key_column="order_id")
print("v2: MERGE (2 updates, 1 insert)")

# Version 3: More inserts
orders_table.insert([
    {"order_id": 5, "customer": "Eve", "amount": 500, "status": "pending"},
])
print("v3: INSERT (1 new order)")

# Show history
print(f"\n📜 TABLE HISTORY:")
for entry in orders_table.history():
    print(f"   {entry}")

# Time travel
print(f"\n⏰ TIME TRAVEL — Query as of Version 1:")
v1_data = orders_table.time_travel(1)
for r in v1_data:
    print(f"   {r}")

print(f"\n⏰ TIME TRAVEL — Query as of Version 3 (current):")
v3_data = orders_table.time_travel(3)
for r in v3_data:
    print(f"   {r}")

print("\n💡 Key Lakehouse Features Demonstrated:")
print("   ✅ ACID transactions (atomic inserts and merges)")
print("   ✅ Time travel (query any historical version)")
print("   ✅ MERGE/Upsert (insert or update in one operation)")
print("   ✅ Audit history (who changed what, when)")
````

## Section 3.4: Batch Processing

Batch processing handles large volumes of data at scheduled intervals.
It's the workhorse of data engineering — most pipelines are batch.

### When to Use Batch vs Streaming

| Factor | Use Batch | Use Streaming |
|--------|-----------|---------------|
| **Latency tolerance** | Minutes to hours OK | Seconds required |
| **Data volume** | Large (GB to TB per run) | Continuous flow |
| **Complexity** | Simpler to build and debug | More complex |
| **Cost** | Lower (scheduled compute) | Higher (always-on) |
| **Use cases** | Reports, ML training, backfills | Alerts, real-time dashboards, fraud |
| **Error handling** | Retry entire batch | Complex (checkpoints, DLQ) |

### Batch Processing Patterns

**1. Full Refresh (Simplest)**
```python
# Delete everything, reload from scratch
target_table.overwrite(source_query.execute())
```
- ✅ Simple, always correct
- ❌ Expensive for large tables, loses history

**2. Incremental Append**
```python
# Only process new records since last run
new_records = source.where(f"updated_at > '{last_run_time}'")
target_table.append(new_records)
```
- ✅ Efficient, fast
- ❌ Doesn't handle updates or deletes

**3. Upsert (MERGE)**
```python
# Insert new records, update existing ones
target_table.merge(
    source=new_and_changed_records,
    condition="target.id = source.id",
    when_matched="UPDATE SET *",
    when_not_matched="INSERT *"
)
```
- ✅ Handles inserts AND updates
- ❌ More complex, needs unique key

**4. Idempotent Processing (Gold Standard)**
```python
# Safe to rerun — same input always produces same output
def process_date(date):
    target_table.delete(f"partition_date = '{date}'")
    target_table.insert(transform(source.where(f"date = '{date}'")))
```
- ✅ Safe to retry, deterministic
- ❌ Slightly more complex

### Production Example: Stripe's Daily Reconciliation

Stripe processes 1B+ financial transactions daily:
- **Schedule**: Every night at 2 AM UTC
- **Pattern**: Idempotent (safe to rerun for any date)
- **Validation**: Every dollar must balance (zero tolerance)
- **Duration**: ~4 hours for full reconciliation
- **Retry logic**: Automatic retry 3x, then page on-call engineer

````python
# Batch processing patterns — idempotent pipeline
from datetime import date, timedelta

class BatchPipeline:
    """Demonstrates idempotent batch processing."""
    
    def __init__(self, name):
        self.name = name
        self.target_data = {}  # Partitioned by date
        self.run_log = []
    
    def run(self, process_date, source_data):
        """
        Idempotent batch run for a specific date.
        Safe to rerun — always produces the same result.
        """
        print(f"🔄 Running pipeline '{self.name}' for {process_date}")
        
        # Step 1: Delete existing data for this date (idempotency)
        if process_date in self.target_data:
            old_count = len(self.target_data[process_date])
            print(f"   ♻️ Clearing existing data ({old_count} records)")
        
        # Step 2: Filter source data for this date
        date_data = [r for r in source_data if r["date"] == str(process_date)]
        print(f"   📥 Source records for {process_date}: {len(date_data)}")
        
        # Step 3: Transform
        transformed = self._transform(date_data)
        print(f"   🔄 Transformed records: {len(transformed)}")
        
        # Step 4: Write to target partition
        self.target_data[process_date] = transformed
        print(f"   ✅ Written to target partition: {process_date}")
        
        # Step 5: Log the run
        self.run_log.append({
            "date": process_date,
            "records": len(transformed),
            "status": "success"
        })
        
        return transformed
    
    def _transform(self, records):
        """Apply business transformations."""
        # Aggregate by customer
        customer_totals = {}
        for r in records:
            cust = r["customer_id"]
            if cust not in customer_totals:
                customer_totals[cust] = {"customer_id": cust, "date": r["date"],
                                         "total_orders": 0, "total_revenue": 0}
            customer_totals[cust]["total_orders"] += 1
            customer_totals[cust]["total_revenue"] += r["amount"]
        return list(customer_totals.values())
    
    def backfill(self, start_date, end_date, source_data):
        """Reprocess a range of dates (backfill)."""
        print(f"\n🔙 BACKFILL: {start_date} to {end_date}")
        current = start_date
        while current <= end_date:
            self.run(current, source_data)
            current += timedelta(days=1)
        print(f"\n✅ Backfill complete: {(end_date - start_date).days + 1} days processed")


# Demo
pipeline = BatchPipeline("daily_customer_revenue")

# Simulate source data
source = [
    {"order_id": 1, "customer_id": "C1", "amount": 100, "date": "2024-03-15"},
    {"order_id": 2, "customer_id": "C1", "amount": 50, "date": "2024-03-15"},
    {"order_id": 3, "customer_id": "C2", "amount": 200, "date": "2024-03-15"},
    {"order_id": 4, "customer_id": "C1", "amount": 75, "date": "2024-03-16"},
    {"order_id": 5, "customer_id": "C3", "amount": 300, "date": "2024-03-16"},
]

# Run for March 15
pipeline.run(date(2024, 3, 15), source)

# Run again (idempotent — same result!)
print("\n--- RERUN (proving idempotency) ---")
pipeline.run(date(2024, 3, 15), source)

# Backfill multiple days
pipeline.backfill(date(2024, 3, 15), date(2024, 3, 16), source)

print("\n💡 Key Principle: IDEMPOTENCY")
print("   Running the pipeline twice produces the SAME result.")
print("   This makes retries safe and backfills trivial.")
````

## Section 3.5: Stream Processing

Stream processing handles data continuously as it arrives, enabling real-time analytics and actions.

### Core Streaming Concepts

| Concept | Definition | Example |
|---------|-----------|---------|
| **Event** | A single data record in the stream | User clicked "Buy" button |
| **Producer** | System that generates events | Mobile app, IoT sensor |
| **Consumer** | System that processes events | Analytics engine, alerting |
| **Topic/Stream** | Named channel for events | "user-clicks", "orders" |
| **Offset** | Position in the stream | "I've processed up to event #42,567" |
| **Watermark** | How late data can arrive | "Accept data up to 5 minutes late" |
| **Window** | Time-based grouping | "Count events per 1-minute window" |
| **Checkpoint** | Save processing state | "If I crash, resume from here" |

### Windowing Types

```
Events: ──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──

TUMBLING WINDOW (fixed, non-overlapping):
|  Window 1  |  Window 2  |  Window 3  |  Window 4  |
|──●──●──●──|──●──●──●──|──●──●──●──|──●──●──●──|

SLIDING WINDOW (fixed, overlapping):
|  Window 1  |
   |  Window 2  |
      |  Window 3  |
         |  Window 4  |

SESSION WINDOW (gap-based, variable size):
|──●●●──|     |──●●●●●●──|     |──●●──|
  (gap)           (gap)           (gap)
```

### The Late Data Problem

In distributed systems, data can arrive out of order:

```
Real time:     10:00  10:01  10:02  10:03  10:04  10:05
Event created: 10:00  10:01  10:02  10:03  10:04  10:05
Event arrives: 10:00  10:01  10:05  10:02  10:03  10:06
                                    ↑       ↑
                              LATE! (3 min)  LATE! (arrived after 10:02)
```

**Solutions:**
- **Watermarks**: "Process events up to 5 minutes late, then close the window"
- **Allowed lateness**: Accept late events and update results
- **Dead letter queue**: Send very late events to a separate queue for manual review

### Production Example: Twitter's Trending Topics

Twitter processes 500M+ tweets/day to determine trending topics:
- **Input**: Tweet stream (text, hashtags, location, timestamp)
- **Processing**: Sliding window (5 min), count hashtag frequency
- **Comparison**: Current frequency vs historical baseline
- **Output**: Topics with abnormally high frequency = "trending"
- **Latency**: New trends detected within 2-3 minutes

````python
# Stream processing simulation with windowing
from datetime import datetime, timedelta
from collections import defaultdict

class StreamProcessor:
    """Simulates a stream processing engine with windowing."""
    
    def __init__(self, window_size_seconds=60, watermark_seconds=30):
        self.window_size = timedelta(seconds=window_size_seconds)
        self.watermark = timedelta(seconds=watermark_seconds)
        self.windows = defaultdict(list)
        self.results = []
        self.late_events = []
    
    def _get_window_start(self, event_time):
        """Determine which window an event belongs to (tumbling)."""
        seconds = int(event_time.timestamp())
        window_seconds = int(self.window_size.total_seconds())
        window_start = seconds - (seconds % window_seconds)
        return datetime.fromtimestamp(window_start)
    
    def process_event(self, event, processing_time=None):
        """Process a single event."""
        event_time = event["event_time"]
        if processing_time is None:
            processing_time = datetime.now()
        
        window_start = self._get_window_start(event_time)
        
        # Check if event is too late (beyond watermark)
        latest_window = max(self.windows.keys()) if self.windows else window_start
        if event_time < latest_window - self.watermark:
            self.late_events.append(event)
            return "LATE"
        
        self.windows[window_start].append(event)
        return "PROCESSED"
    
    def emit_results(self):
        """Emit aggregated results for closed windows."""
        for window_start, events in sorted(self.windows.items()):
            window_end = window_start + self.window_size
            result = {
                "window_start": window_start.strftime("%H:%M:%S"),
                "window_end": window_end.strftime("%H:%M:%S"),
                "event_count": len(events),
                "unique_users": len(set(e.get("user_id", "unknown") for e in events)),
            }
            self.results.append(result)
        return self.results


# Simulate a clickstream
base_time = datetime(2024, 3, 15, 10, 0, 0)
events = [
    {"event_time": base_time + timedelta(seconds=5), "user_id": "u1", "action": "click"},
    {"event_time": base_time + timedelta(seconds=15), "user_id": "u2", "action": "view"},
    {"event_time": base_time + timedelta(seconds=30), "user_id": "u1", "action": "purchase"},
    {"event_time": base_time + timedelta(seconds=45), "user_id": "u3", "action": "click"},
    {"event_time": base_time + timedelta(seconds=65), "user_id": "u1", "action": "view"},  # Window 2
    {"event_time": base_time + timedelta(seconds=80), "user_id": "u2", "action": "click"},  # Window 2
    {"event_time": base_time + timedelta(seconds=125), "user_id": "u4", "action": "view"},  # Window 3
    # Late event (belongs to window 1 but arrives during window 3)
    {"event_time": base_time + timedelta(seconds=10), "user_id": "u5", "action": "click"},
]

processor = StreamProcessor(window_size_seconds=60, watermark_seconds=30)

print("🌊 STREAM PROCESSING SIMULATION")
print("=" * 60)
print(f"   Window size: 60 seconds (tumbling)")
print(f"   Watermark: 30 seconds (accept late data up to 30s)")
print()

for i, event in enumerate(events):
    status = processor.process_event(event)
    time_str = event["event_time"].strftime("%H:%M:%S")
    print(f"   Event {i+1}: time={time_str}, user={event['user_id']}, "
          f"action={event['action']} → {status}")

print(f"\n📊 WINDOW RESULTS:")
results = processor.emit_results()
for r in results:
    print(f"   [{r['window_start']} - {r['window_end']}] "
          f"events={r['event_count']}, unique_users={r['unique_users']}")

print(f"\n⚠️ Late events (beyond watermark): {len(processor.late_events)}")
for e in processor.late_events:
    print(f"   {e['event_time'].strftime('%H:%M:%S')}: {e['user_id']} - {e['action']}")

print("\n💡 Key Streaming Concepts Demonstrated:")
print("   • Tumbling windows group events by fixed time intervals")
print("   • Watermarks define how late data can arrive")
print("   • Late events are handled separately (DLQ or dropped)")
````

## Section 3.8: Performance Optimization

Performance optimization is about doing more with less — faster queries, lower costs, better throughput.

### The Performance Optimization Hierarchy

```
1. DATA MODELING (biggest impact)
   └── Choose the right model (star schema vs OBT vs normalized)
   └── Denormalize for read-heavy workloads

2. STORAGE OPTIMIZATION
   └── Partitioning (filter entire file groups)
   └── Clustering/Z-ordering (co-locate related data)
   └── File sizing (avoid small files problem)
   └── Compression (reduce I/O)

3. QUERY OPTIMIZATION
   └── Predicate pushdown (filter early)
   └── Column pruning (read only needed columns)
   └── Broadcast joins (small table fits in memory)
   └── Avoid shuffles (expensive network I/O)

4. COMPUTE OPTIMIZATION
   └── Right-size clusters
   └── Auto-scaling
   └── Caching hot data
   └── Spot/preemptible instances
```

### Key Optimization Techniques

| Technique | What It Does | Impact |
|-----------|-------------|--------|
| **Partitioning** | Organize data by a column (date, region) | Skip irrelevant files (10-100x faster) |
| **Clustering** | Sort data within partitions | Reduce data scanned (2-10x faster) |
| **Caching** | Keep hot data in memory | Eliminate repeated I/O (10-100x faster) |
| **Predicate Pushdown** | Filter at storage level | Read less data (2-50x faster) |
| **Broadcast Join** | Send small table to all nodes | Avoid shuffle (5-20x faster) |
| **AQE** | Adapt query plan at runtime | Auto-optimize (1.5-5x faster) |

# 📗 PART 7: Best Practices & Design Principles

## Section 7.1: The Seven Principles of Data Engineering

### 1. Idempotency — Safe to Rerun

> "If I run this pipeline twice, do I get the same result?"

An idempotent pipeline produces the same output regardless of how many times it runs.
This is the MOST IMPORTANT principle in data engineering.

**Why it matters:**
- Pipelines fail (network issues, OOM, timeouts)
- You WILL need to retry
- Without idempotency, retries create duplicates or corruption

**How to achieve it:**
- Use MERGE instead of INSERT (upsert on unique key)
- Use partition overwrite (delete + insert for a date partition)
- Use deterministic processing (same input → same output)

### 2. Immutability — Never Modify Raw Data

> "Raw data is sacred. Transform it, don't mutate it."

**Pattern:**
```
raw_data (NEVER TOUCH) → cleaned_data → enriched_data → served_data
```

**Why:** If your transformation has a bug, you can always go back to raw and reprocess.

### 3. Incremental Processing — Only Process New Data

> "Don't boil the ocean every day."

Process only what's changed since the last run. This reduces:
- Compute costs (process 1% instead of 100%)
- Runtime (minutes instead of hours)
- Resource contention

### 4. Schema Evolution — Handle Changes Gracefully

> "Source systems WILL change their schema. Be ready."

**Strategies:**
- Schema-on-read (flexible, but risky)
- Schema enforcement with evolution (add columns OK, remove = error)
- Schema registry (centralized schema management)

### 5. Separation of Concerns — Decouple Stages

> "Ingestion shouldn't know about transformation. Transformation shouldn't know about serving."

Each stage should be independently deployable, testable, and scalable.

### 6. Fail Fast — Detect Errors Early

> "A data quality issue caught in Bronze costs $1 to fix. Caught in Gold, it costs $1000."

Validate data at every stage boundary. Don't let bad data propagate.

### 7. Observability — Monitor Everything

> "If it's not monitored, it's not in production."

Monitor: pipeline success/failure, data freshness, quality scores, costs, latency.

````python
# Demonstrating idempotent pipeline design
class IdempotentPipeline:
    """
    A production-quality idempotent pipeline.
    Safe to rerun any number of times — always produces the same result.
    """
    
    def __init__(self, name, target_table):
        self.name = name
        self.target_table = target_table
        self.data = {}  # Simulated table (partition → records)
        self.run_count = 0
    
    def run(self, partition_key, source_records):
        """
        Idempotent run: overwrite the partition, don't append.
        Running this 1x or 100x produces the same result.
        """
        self.run_count += 1
        
        # Step 1: Transform (deterministic — no random, no timestamps)
        transformed = self._transform(source_records)
        
        # Step 2: Overwrite partition (idempotent write)
        self.data[partition_key] = transformed  # Replace, don't append!
        
        return len(transformed)
    
    def _transform(self, records):
        """Deterministic transformation — same input always gives same output."""
        result = []
        for r in records:
            result.append({
                "id": r["id"],
                "value": r["amount"] * 1.1,  # Add 10% tax
                "category": r["category"].upper(),
            })
        return sorted(result, key=lambda x: x["id"])  # Deterministic order
    
    def get_data(self):
        """Get all data across all partitions."""
        all_records = []
        for partition, records in sorted(self.data.items()):
            all_records.extend(records)
        return all_records


# Prove idempotency
pipeline = IdempotentPipeline("daily_sales", "gold.daily_sales")

source = [
    {"id": 1, "amount": 100, "category": "electronics"},
    {"id": 2, "amount": 200, "category": "clothing"},
    {"id": 3, "amount": 300, "category": "food"},
]

print("🔄 IDEMPOTENCY PROOF")
print("=" * 60)

# Run 1
pipeline.run("2024-03-15", source)
result_1 = pipeline.get_data()
print(f"Run 1: {len(result_1)} records")
for r in result_1:
    print(f"   {r}")

# Run 2 (same input — should produce IDENTICAL output)
pipeline.run("2024-03-15", source)
result_2 = pipeline.get_data()
print(f"\nRun 2: {len(result_2)} records")
for r in result_2:
    print(f"   {r}")

# Run 3 (third time — still identical)
pipeline.run("2024-03-15", source)
result_3 = pipeline.get_data()
print(f"\nRun 3: {len(result_3)} records")

# Verify
assert result_1 == result_2 == result_3, "NOT IDEMPOTENT!"
print(f"\n✅ VERIFIED: All 3 runs produced IDENTICAL results")
print(f"   Records after run 1: {len(result_1)}")
print(f"   Records after run 2: {len(result_2)} (no duplicates!)")
print(f"   Records after run 3: {len(result_3)} (still no duplicates!)")
print(f"\n💡 Key: We OVERWRITE the partition, not APPEND to it.")
print(f"   This makes retries safe and backfills trivial.")

# Contrast with NON-idempotent (dangerous!)
print("\n\n❌ NON-IDEMPOTENT EXAMPLE (what NOT to do):")
bad_data = []
for run in range(3):
    bad_data.extend(source)  # APPEND every time!
print(f"   After 3 runs with APPEND: {len(bad_data)} records (should be 3, got 9!)")
print(f"   ⚠️ This creates DUPLICATES on every retry!")
````

### Pattern Comparison

### Error Handling Patterns

**1. Retry with Exponential Backoff**
```python
def retry_with_backoff(func, max_retries=3):
    for attempt in range(max_retries):
        try:
            return func()
        except TransientError:
            wait = 2 ** attempt  # 1s, 2s, 4s, 8s...
            time.sleep(wait)
    raise MaxRetriesExceeded()
```

**2. Dead Letter Queue (DLQ)**
```
Normal flow:  Event → Process → Success → Target
Error flow:   Event → Process → FAIL → Dead Letter Queue → Manual review
```

**3. Circuit Breaker**
```
Normal:  Requests flow through
Open:    Too many failures → stop sending requests (protect downstream)
Half-open: Try one request → if success, close circuit
```

````python
# Error handling patterns for production pipelines
import time
import random

class PipelineError(Exception):
    pass

class TransientError(PipelineError):
    """Temporary error — safe to retry (network timeout, rate limit)."""
    pass

class PermanentError(PipelineError):
    """Permanent error — do NOT retry (bad data, schema mismatch)."""
    pass

def retry_with_exponential_backoff(func, max_retries=3, base_delay=1):
    """
    Retry a function with exponential backoff.
    
    Delays: 1s, 2s, 4s, 8s, 16s...
    Only retries on TransientError (not PermanentError).
    """
    for attempt in range(max_retries + 1):
        try:
            result = func()
            if attempt > 0:
                print(f"   ✅ Succeeded on attempt {attempt + 1}")
            return result
        except TransientError as e:
            if attempt == max_retries:
                print(f"   ❌ Max retries ({max_retries}) exceeded. Giving up.")
                raise
            delay = base_delay * (2 ** attempt)
            print(f"   ⚠️ Attempt {attempt + 1} failed: {e}")
            print(f"      Retrying in {delay}s (exponential backoff)...")
            # time.sleep(delay)  # Commented out for demo speed
        except PermanentError as e:
            print(f"   ❌ Permanent error (no retry): {e}")
            raise

class DeadLetterQueue:
    """Dead Letter Queue for failed records."""
    
    def __init__(self, name):
        self.name = name
        self.messages = []
    
    def send(self, record, error, timestamp=None):
        self.messages.append({
            "record": record,
            "error": str(error),
            "timestamp": timestamp or "2024-03-15T10:00:00Z",
            "retry_count": 0
        })
    
    def get_pending(self):
        return [m for m in self.messages if m["retry_count"] < 3]

class CircuitBreaker:
    """Circuit breaker pattern — protect downstream services."""
    
    def __init__(self, failure_threshold=3, reset_timeout=60):
        self.failure_threshold = failure_threshold
        self.reset_timeout = reset_timeout
        self.failure_count = 0
        self.state = "CLOSED"  # CLOSED (normal), OPEN (blocking), HALF_OPEN (testing)
        self.last_failure_time = None
    
    def call(self, func):
        if self.state == "OPEN":
            print(f"   🚫 Circuit OPEN — request blocked (protecting downstream)")
            raise PipelineError("Circuit breaker is OPEN")
        
        try:
            result = func()
            self.failure_count = 0
            if self.state == "HALF_OPEN":
                self.state = "CLOSED"
                print(f"   🟢 Circuit CLOSED (recovered)")
            return result
        except Exception as e:
            self.failure_count += 1
            if self.failure_count >= self.failure_threshold:
                self.state = "OPEN"
                print(f"   🔴 Circuit OPEN (too many failures: {self.failure_count})")
            raise


# Demo: Retry with backoff
print("🔄 RETRY WITH EXPONENTIAL BACKOFF")
print("=" * 50)
attempt_counter = {"count": 0}

def flaky_api_call():
    """Simulates an API that fails twice then succeeds."""
    attempt_counter["count"] += 1
    if attempt_counter["count"] <= 2:
        raise TransientError("Connection timeout")
    return {"status": "success", "data": [1, 2, 3]}

result = retry_with_exponential_backoff(flaky_api_call, max_retries=3)
print(f"   Result: {result}")

# Demo: Dead Letter Queue
print("\n\n📬 DEAD LETTER QUEUE")
print("=" * 50)
dlq = DeadLetterQueue("failed_orders")

# Process records, send failures to DLQ
records = [
    {"order_id": 1, "amount": 100},
    {"order_id": 2, "amount": -50},  # Invalid!
    {"order_id": 3, "amount": 200},
    {"order_id": 4, "amount": None},  # Invalid!
]

for record in records:
    try:
        if record["amount"] is None or record["amount"] < 0:
            raise PermanentError(f"Invalid amount: {record['amount']}")
        print(f"   ✅ Processed order {record['order_id']}")
    except PermanentError as e:
        dlq.send(record, e)
        print(f"   ❌ Order {record['order_id']} → DLQ: {e}")

print(f"\n   DLQ size: {len(dlq.messages)} messages pending review")

# Demo: Circuit Breaker
print("\n\n🔌 CIRCUIT BREAKER")
print("=" * 50)
cb = CircuitBreaker(failure_threshold=3)
call_count = {"n": 0}

def unreliable_service():
    call_count["n"] += 1
    if call_count["n"] <= 4:
        raise TransientError("Service unavailable")
    return "OK"

for i in range(5):
    try:
        cb.call(unreliable_service)
        print(f"   Call {i+1}: ✅ Success")
    except Exception as e:
        print(f"   Call {i+1}: ❌ {e}")
````

````python
# Skills assessment and learning plan generator
def generate_learning_plan(current_skills, target_role, available_hours_per_week):
    """Generate a personalized learning plan based on skill gaps."""
    
    role_requirements = {
        "junior_de": {
            "SQL": 3, "Python": 2, "Cloud": 1, "Spark": 1,
            "Data Modeling": 1, "Git": 2, "ETL": 2
        },
        "mid_de": {
            "SQL": 4, "Python": 3, "Cloud": 3, "Spark": 3,
            "Data Modeling": 3, "Git": 3, "ETL": 4, "Orchestration": 3,
            "Data Quality": 2, "Streaming": 2
        },
        "senior_de": {
            "SQL": 5, "Python": 4, "Cloud": 4, "Spark": 4,
            "Data Modeling": 4, "Git": 4, "ETL": 5, "Orchestration": 4,
            "Data Quality": 4, "Streaming": 3, "System Design": 4,
            "Leadership": 3, "CI/CD": 3
        }
    }
    
    requirements = role_requirements.get(target_role, role_requirements["mid_de"])
    
    # Calculate gaps
    gaps = {}
    for skill, required_level in requirements.items():
        current = current_skills.get(skill, 0)
        if current < required_level:
            gaps[skill] = {
                "current": current,
                "required": required_level,
                "gap": required_level - current,
                "hours_to_close": (required_level - current) * 20  # ~20 hours per level
            }
    
    # Sort by gap size (biggest gaps first)
    sorted_gaps = sorted(gaps.items(), key=lambda x: x[1]["gap"], reverse=True)
    
    total_hours = sum(g["hours_to_close"] for _, g in sorted_gaps)
    weeks_needed = total_hours / available_hours_per_week
    
    print(f"📋 PERSONALIZED LEARNING PLAN")
    print(f"   Target role: {target_role.replace('_', ' ').title()}")
    print(f"   Available time: {available_hours_per_week} hours/week")
    print(f"   Total hours needed: {total_hours}")
    print(f"   Estimated timeline: {weeks_needed:.0f} weeks ({weeks_needed/4:.0f} months)")
    print("=" * 60)
    
    if not sorted_gaps:
        print("   ✅ No skill gaps! You're ready for this role.")
        return
    
    print(f"\n{'Skill':<18} {'Current':<10} {'Required':<10} {'Gap':<6} {'Hours'}")
    print("-" * 60)
    for skill, details in sorted_gaps:
        bar = "░" * details["current"] + "█" * details["gap"]
        print(f"{skill:<18} {details['current']}/5{'░'*details['current']:<7} "
              f"{details['required']}/5{'█'*details['required']:<7} "
              f"{details['gap']:<6} ~{details['hours_to_close']}h")
    
    # Generate weekly plan
    print(f"\n📅 SUGGESTED WEEKLY SCHEDULE ({available_hours_per_week}h/week):")
    week = 1
    for skill, details in sorted_gaps[:5]:  # Top 5 priorities
        weeks_for_skill = details["hours_to_close"] / available_hours_per_week
        print(f"   Weeks {week}-{week + int(weeks_for_skill)}: {skill} "
              f"(level {details['current']} → {details['required']})")
        week += int(weeks_for_skill) + 1


# Example: Someone transitioning from analyst to data engineer
print("EXAMPLE: Analyst → Mid-Level Data Engineer")
print()
generate_learning_plan(
    current_skills={
        "SQL": 3,      # Good at SQL from analyst work
        "Python": 1,   # Basic scripting
        "Cloud": 0,    # No cloud experience
        "Spark": 0,    # Never used
        "Data Modeling": 2,  # Some dimensional modeling
        "Git": 1,      # Basic commits
        "ETL": 1,      # Used SSIS once
    },
    target_role="mid_de",
    available_hours_per_week=10
)
````

````python
# ============================================
# ✅ SOLUTION: Schema Evolution Strategy
# ============================================
schema_evolution_strategy = {
    "detection": {
        "method": "Compare source schema vs expected schema on every pipeline run",
        "implementation": """
            1. Store expected schema in a schema registry (JSON/Avro)
            2. On each run, fetch current source schema
            3. Compare: new columns? removed columns? type changes?
            4. Act based on change type (see below)
        """,
        "tools": ["Schema Registry (Confluent)", "Great Expectations", "Custom Python"]
    },
    
    "handling_rules": {
        "new_column_added": {
            "action": "AUTO-ADD to Bronze, ALERT for Silver/Gold",
            "reason": "Bronze should capture everything. Silver/Gold need review.",
            "implementation": "mergeSchema=True for Bronze, strict for Silver"
        },
        "column_removed": {
            "action": "ALERT immediately, do NOT remove from target",
            "reason": "Downstream consumers may depend on this column.",
            "implementation": "Fill with NULL, alert data owner"
        },
        "type_changed": {
            "action": "BLOCK pipeline, require manual review",
            "reason": "Type changes can break downstream queries silently.",
            "implementation": "Schema validation check, page on-call"
        },
        "column_renamed": {
            "action": "ALERT, map old name → new name in config",
            "reason": "Renaming is ambiguous (is it a rename or remove+add?)",
            "implementation": "Column mapping configuration file"
        }
    },
    
    "prevention": [
        "1. Data contracts with source teams (14-day notice for breaking changes)",
        "2. Schema registry with compatibility checks (backward/forward)",
        "3. Integration tests that validate schema before production deploy",
        "4. Monitoring: alert on any schema drift within 5 minutes",
        "5. Documentation: maintain a schema changelog"
    ],
    
    "communication": [
        "Slack alert to #data-engineering channel",
        "Auto-create Jira ticket for schema review",
        "Update data catalog with new/changed columns",
        "Notify downstream consumers via email/Slack"
    ]
}

print("🔄 SCHEMA EVOLUTION STRATEGY")
print("=" * 60)

print("\n📋 DETECTION:")
print(f"   Method: {schema_evolution_strategy['detection']['method']}")

print("\n📋 HANDLING RULES:")
for change_type, rule in schema_evolution_strategy["handling_rules"].items():
    print(f"\n   {change_type}:")
    print(f"      Action: {rule['action']}")
    print(f"      Reason: {rule['reason']}")

print("\n📋 PREVENTION MEASURES:")
for measure in schema_evolution_strategy["prevention"]:
    print(f"   {measure}")

print("\n📋 COMMUNICATION:")
for comm in schema_evolution_strategy["communication"]:
    print(f"   • {comm}")

print("\n💡 Key Principle: Schema changes should NEVER break pipelines silently.")
print("   Detect early, handle gracefully, communicate immediately.")
````

## Exercise 4: Design for Failure

**Scenario:** Your pipeline processes 1M records/hour from a Kafka topic.
At 2:47 AM, the pipeline crashes halfway through processing hour 02:00-03:00.
500K records were processed, 500K were not.

**Your task:** Design the pipeline so this failure is handled gracefully.

````python
# ============================================
# 🎯 YOUR TURN: Design for Failure
# ============================================
# Consider:
# 1. How do you know where you left off? (checkpointing)
# 2. How do you avoid duplicates when you restart? (idempotency)
# 3. How do you alert the team? (monitoring)
# 4. How do you auto-recover? (retry logic)
# 5. What if the failure keeps happening? (circuit breaker)
#
# Write your failure-handling design below:
````

````python
# ============================================
# ✅ SOLUTION: Failure-Resilient Pipeline Design
# ============================================
failure_resilient_design = {
    "checkpointing": {
        "what": "Save processing progress to durable storage",
        "how": "After each micro-batch, save the Kafka offset to Delta table",
        "benefit": "On restart, resume from last successful offset (no reprocessing)",
        "implementation": """
            # Spark Structured Streaming does this automatically:
            stream.writeStream
                .option("checkpointLocation", "/checkpoints/my_pipeline")
                .start()
            # On restart, reads checkpoint and resumes from last offset
        """
    },
    
    "idempotency": {
        "what": "Ensure reprocessing doesn't create duplicates",
        "how": "Use MERGE with deduplication on (event_id, event_timestamp)",
        "benefit": "Safe to reprocess any time window without side effects",
        "implementation": """
            # MERGE ensures no duplicates:
            MERGE INTO target USING source
            ON target.event_id = source.event_id
            WHEN NOT MATCHED THEN INSERT *
            # Even if we process the same events twice, no duplicates!
        """
    },
    
    "monitoring_and_alerting": {
        "what": "Detect failures immediately and notify the team",
        "how": "Monitor pipeline lag, error rate, and throughput",
        "alerts": [
            "Pipeline stopped for > 5 minutes → PagerDuty (on-call)",
            "Consumer lag > 100K events → Slack warning",
            "Error rate > 1% → Slack alert",
            "No data processed for > 15 minutes → PagerDuty critical"
        ]
    },
    
    "auto_recovery": {
        "what": "Automatically retry on transient failures",
        "how": "Exponential backoff with max 3 retries",
        "escalation": [
            "Retry 1: Wait 30 seconds, restart pipeline",
            "Retry 2: Wait 2 minutes, restart with more memory",
            "Retry 3: Wait 5 minutes, restart on different node",
            "All retries failed: Page on-call engineer"
        ]
    },
    
    "dead_letter_queue": {
        "what": "Isolate poison pill records that always fail",
        "how": "After 3 failures for a specific record, send to DLQ",
        "benefit": "One bad record doesn't block the entire pipeline",
        "review": "DLQ reviewed daily by data quality team"
    }
}

print("🛡️ FAILURE-RESILIENT PIPELINE DESIGN")
print("=" * 60)
for component, details in failure_resilient_design.items():
    print(f"\n📌 {component.upper().replace('_', ' ')}:")
    print(f"   What: {details['what']}")
    print(f"   How: {details['how']}")
    if "benefit" in details:
        print(f"   Benefit: {details['benefit']}")
    if "alerts" in details:
        print(f"   Alerts:")
        for alert in details["alerts"]:
            print(f"      🚨 {alert}")
    if "escalation" in details:
        print(f"   Escalation:")
        for step in details["escalation"]:
            print(f"      → {step}")

print("\n\n💡 THE FAILURE SCENARIO RESOLVED:")
print("   2:47 AM: Pipeline crashes (500K of 1M processed)")
print("   2:47 AM: Monitoring detects stoppage, alerts on-call")
print("   2:48 AM: Auto-retry #1 kicks in (30s delay)")
print("   2:48 AM: Pipeline restarts from checkpoint (offset 500,001)")
print("   2:48 AM: Processes remaining 500K records")
print("   2:55 AM: Pipeline caught up, no data loss, no duplicates")
print("   2:55 AM: Auto-recovery alert: 'Pipeline recovered successfully'")
print("\n   Total impact: 8 minutes of delay. Zero data loss. Zero duplicates.")
````

## Exercise 5: Cost Optimization

**Scenario:** Your data warehouse bill is $50K/month and growing 20% quarterly.
The CFO wants you to reduce it by 50% without losing functionality.

**Your task:** Identify cost drivers and propose optimizations.

````python
# ============================================
# 🎯 YOUR TURN: Reduce Data Platform Costs by 50%
# ============================================
# Current monthly costs:
# - Compute (Spark clusters): $25K
# - Storage (S3/Delta): $10K
# - Data warehouse queries: $10K
# - Networking/egress: $3K
# - Monitoring/tools: $2K
# Total: $50K/month
#
# For each cost category, propose at least 2 optimizations.
# Calculate expected savings for each.
#
# Write your cost optimization plan below:
````

````python
# ============================================
# ✅ SOLUTION: Cost Optimization Plan
# ============================================
cost_optimization = {
    "current_costs": {
        "Compute (Spark)": 25000,
        "Storage (S3/Delta)": 10000,
        "Warehouse queries": 10000,
        "Networking": 3000,
        "Monitoring/tools": 2000,
    },
    "optimizations": {
        "Compute (Spark) — $25K → $10K": [
            {
                "action": "Use spot/preemptible instances for batch jobs",
                "savings": "$10K (70% discount on 60% of compute)",
                "risk": "Jobs may be interrupted (mitigate with checkpointing)"
            },
            {
                "action": "Right-size clusters (most are 3x over-provisioned)",
                "savings": "$5K (reduce from 10 nodes to 4 for most jobs)",
                "risk": "Low (monitor and adjust)"
            },
            {
                "action": "Auto-terminate idle clusters after 15 minutes",
                "savings": "$3K (developers leave clusters running overnight)",
                "risk": "None"
            }
        ],
        "Storage (S3/Delta) — $10K → $5K": [
            {
                "action": "Lifecycle policies: move cold data to S3 Glacier",
                "savings": "$3K (data older than 90 days → 80% cheaper storage)",
                "risk": "Slower access to historical data (acceptable)"
            },
            {
                "action": "VACUUM old Delta versions (keep 30 days, not forever)",
                "savings": "$1.5K (remove old file versions)",
                "risk": "Lose time travel beyond 30 days"
            },
            {
                "action": "Better compression (switch from Snappy to ZSTD)",
                "savings": "$0.5K (20% smaller files)",
                "risk": "Slightly slower writes (negligible)"
            }
        ],
        "Warehouse queries — $10K → $4K": [
            {
                "action": "Optimize top 20 most expensive queries",
                "savings": "$3K (top 20 queries = 60% of cost)",
                "risk": "None (just better SQL)"
            },
            {
                "action": "Pre-aggregate common queries into materialized views",
                "savings": "$2K (avoid repeated full scans)",
                "risk": "Slight staleness (refresh every hour)"
            },
            {
                "action": "Partition hot tables by date",
                "savings": "$1K (scan less data per query)",
                "risk": "None"
            }
        ],
        "Networking — $3K → $1.5K": [
            {
                "action": "Keep compute and storage in same region",
                "savings": "$1K (eliminate cross-region transfer)",
                "risk": "None"
            },
            {
                "action": "Compress data before transfer",
                "savings": "$0.5K",
                "risk": "None"
            }
        ]
    }
}

current_total = sum(cost_optimization["current_costs"].values())
print("💰 COST OPTIMIZATION PLAN")
print("=" * 60)
print(f"   Current monthly cost: ${current_total:,}")
print(f"   Target: ${current_total * 0.5:,.0f} (50% reduction)")
print()

total_savings = 0
for category, actions in cost_optimization["optimizations"].items():
    print(f"\n📌 {category}")
    for action in actions:
        print(f"   ✂️ {action['action']}")
        print(f"      Savings: {action['savings']}")
        # Extract numeric savings
        import re
        match = re.search(r'\$([\d.]+)K', action['savings'])
        if match:
            total_savings += float(match.group(1)) * 1000

print(f"\n{'=' * 60}")
print(f"   TOTAL ESTIMATED SAVINGS: ${total_savings:,.0f}/month")
print(f"   New monthly cost: ${current_total - total_savings:,.0f}")
print(f"   Reduction: {total_savings/current_total*100:.0f}%")
print(f"   Annual savings: ${total_savings * 12:,.0f}")
````
