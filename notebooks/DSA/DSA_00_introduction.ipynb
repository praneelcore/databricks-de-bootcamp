# Section 1: What is DSA?

## The Simple Explanation

**Data Structures** = Ways to organize and store data
**Algorithms** = Step-by-step instructions to solve a problem

Think of it like a kitchen:
- **Data structures** are your storage containers (bowls, pots, shelves, fridge)
- **Algorithms** are your recipes (step-by-step instructions to cook a dish)

A good chef knows which container to use for which ingredient, and which recipe
is fastest for the occasion. A good programmer knows which data structure to use
for which problem, and which algorithm is most efficient.

## Real-World Analogy

Imagine you have 1,000 books and need to find one specific book.

**Approach 1 (no structure):** Books scattered randomly on the floor.
You check each book one by one. Average: 500 checks. Worst case: 1,000 checks.

**Approach 2 (sorted + binary search):** Books sorted alphabetically on shelves.
You open to the middle, go left or right, repeat. Maximum: 10 checks!

That's the power of DSA -- the right structure + the right algorithm = dramatically faster.

## What DSA Is NOT

- It's NOT about memorizing code
- It's NOT just for competitive programming
- It's NOT only for "genius" programmers
- It's NOT irrelevant to data engineering

## What DSA IS

- A toolkit for solving problems efficiently
- A way to think about trade-offs (speed vs memory)
- The foundation of every software system you use
- What interviewers test at Google, Meta, Amazon, Databricks

````python
# The power of the right approach -- a concrete example
import time

# Problem: Find if a number exists in a collection of 1 million numbers

# Approach 1: List (unsorted) -- check every element
def search_list(data, target):
    for item in data:
        if item == target:
            return True
    return False

# Approach 2: Set (hash table) -- instant lookup
def search_set(data, target):
    return target in data

# Create 1 million numbers
import random
random.seed(42)
numbers = list(range(1_000_000))
numbers_set = set(numbers)
target = 999_999  # Worst case -- near the end

# Time both approaches
print("Finding 999,999 in 1,000,000 numbers:")
print("=" * 50)

start = time.perf_counter()
result = search_list(numbers, target)
list_time = time.perf_counter() - start
print(f"List search:  {list_time:.6f} seconds  (found: {result})")

start = time.perf_counter()
result = search_set(numbers_set, target)
set_time = time.perf_counter() - start
print(f"Set lookup:   {set_time:.6f} seconds  (found: {result})")

speedup = list_time / set_time
print(f"\nSpeedup: {speedup:,.0f}x faster with the right data structure!")
print("\nThis is why DSA matters. Same problem, same result, wildly different speed.")
````

# Section 2: The Core Data Structures

## Overview

There are about 10 fundamental data structures. Everything else is built from these.

```
DATA STRUCTURES
├── Linear (elements in a sequence)
│   ├── Array / List      -- indexed, fast access by position
│   ├── Linked List       -- nodes connected by pointers
│   ├── Stack             -- Last In, First Out (LIFO)
│   └── Queue             -- First In, First Out (FIFO)
│
├── Non-Linear (elements with relationships)
│   ├── Tree              -- hierarchical (parent-child)
│   │   ├── Binary Tree
│   │   ├── Binary Search Tree (BST)
│   │   └── Heap (priority queue)
│   └── Graph             -- nodes connected by edges
│
└── Hash-Based
    ├── Hash Map / Dict   -- key-value pairs, O(1) lookup
    └── Hash Set          -- unique values, O(1) lookup
```

## When to Use Each

| Data Structure | Use When | Python Equivalent |
|---------------|----------|-------------------|
| **Array/List** | Ordered data, index access | `list` |
| **Hash Map** | Fast key-value lookup | `dict` |
| **Hash Set** | Fast membership check, unique values | `set` |
| **Stack** | Undo/redo, function calls, parsing | `list` (append/pop) |
| **Queue** | Task scheduling, BFS | `collections.deque` |
| **Heap** | Always need min/max quickly | `heapq` module |
| **Tree** | Hierarchical data, sorted data | Custom class |
| **Graph** | Networks, relationships | `dict` of lists |

````python
# All core data structures in Python
from collections import deque
import heapq

print("CORE DATA STRUCTURES IN PYTHON")
print("=" * 60)

# 1. List (Array)
print("\n1. LIST (Array) -- ordered, indexed")
my_list = [10, 20, 30, 40, 50]
print(f"   Data: {my_list}")
print(f"   Access by index: my_list[2] = {my_list[2]}")
print(f"   Append: O(1) amortized")
print(f"   Search: O(n) -- must check each element")

# 2. Dictionary (Hash Map)
print("\n2. DICT (Hash Map) -- key-value pairs")
my_dict = {"alice": 95, "bob": 87, "charlie": 92}
print(f"   Data: {my_dict}")
print(f"   Lookup: my_dict['alice'] = {my_dict['alice']}")
print(f"   Lookup: O(1) average -- instant regardless of size")

# 3. Set (Hash Set)
print("\n3. SET (Hash Set) -- unique values")
my_set = {1, 2, 3, 4, 5}
print(f"   Data: {my_set}")
print(f"   Membership: 3 in my_set = {3 in my_set}")
print(f"   Membership: O(1) average -- instant")

# 4. Stack (using list)
print("\n4. STACK -- Last In, First Out (LIFO)")
stack = []
stack.append("first")
stack.append("second")
stack.append("third")
print(f"   After pushing 3 items: {stack}")
top = stack.pop()
print(f"   After pop: removed '{top}', stack = {stack}")
print(f"   Use case: undo/redo, function call stack, bracket matching")

# 5. Queue (using deque)
print("\n5. QUEUE -- First In, First Out (FIFO)")
queue = deque()
queue.append("first")
queue.append("second")
queue.append("third")
print(f"   After enqueuing 3 items: {list(queue)}")
front = queue.popleft()
print(f"   After dequeue: removed '{front}', queue = {list(queue)}")
print(f"   Use case: task scheduling, BFS, print queue")

# 6. Heap (Priority Queue)
print("\n6. HEAP (Priority Queue) -- always get min/max quickly")
heap = [5, 3, 8, 1, 9, 2]
heapq.heapify(heap)
print(f"   After heapify: {heap}")
print(f"   Get minimum: {heap[0]} (always at index 0)")
smallest = heapq.heappop(heap)
print(f"   After pop: removed {smallest}, heap = {heap}")
print(f"   Use case: Dijkstra's algorithm, top-K problems, scheduling")
````

# Section 3: The Core Algorithms

## What is an Algorithm?

An algorithm is a **finite set of well-defined instructions** to solve a problem.

Key properties:
- **Input**: Takes some data
- **Output**: Produces a result
- **Definiteness**: Each step is clear and unambiguous
- **Finiteness**: It terminates (doesn't run forever)
- **Effectiveness**: Each step is doable

## Algorithm Categories

| Category | What It Does | Examples |
|----------|-------------|---------|
| **Searching** | Find an element | Linear search, Binary search |
| **Sorting** | Order elements | Bubble sort, Merge sort, Quick sort |
| **Traversal** | Visit all elements | BFS, DFS |
| **Divide & Conquer** | Split problem, solve parts | Merge sort, Binary search |
| **Dynamic Programming** | Cache subproblem results | Fibonacci, Knapsack |
| **Greedy** | Make locally optimal choices | Dijkstra's, Huffman coding |
| **Backtracking** | Try all possibilities | N-Queens, Sudoku solver |

## The Most Important Insight

> "There is usually more than one way to solve a problem.
> The question is: which way is FAST ENOUGH for your constraints?"

A solution that works for 100 items might be completely unusable for 1,000,000 items.
This is why we study complexity (Big-O notation) -- to predict how solutions scale.

````python
# Algorithm comparison -- same problem, different approaches
print("ALGORITHM COMPARISON: Finding duplicates in a list")
print("=" * 60)

import time

def has_duplicate_naive(arr):
    """Approach 1: Check every pair -- O(n^2)"""
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                return True
    return False

def has_duplicate_sort(arr):
    """Approach 2: Sort then check neighbors -- O(n log n)"""
    sorted_arr = sorted(arr)
    for i in range(len(sorted_arr) - 1):
        if sorted_arr[i] == sorted_arr[i + 1]:
            return True
    return False

def has_duplicate_set(arr):
    """Approach 3: Use a set -- O(n)"""
    seen = set()
    for x in arr:
        if x in seen:
            return True
        seen.add(x)
    return False

# Test with different sizes
import random
random.seed(42)

print(f"{'Size':<10} {'O(n^2)':<15} {'O(n log n)':<15} {'O(n)':<15}")
print("-" * 55)

for size in [100, 1000, 5000, 10000]:
    arr = list(range(size))  # No duplicates -- worst case for all

    start = time.perf_counter()
    has_duplicate_naive(arr)
    t1 = time.perf_counter() - start

    start = time.perf_counter()
    has_duplicate_sort(arr)
    t2 = time.perf_counter() - start

    start = time.perf_counter()
    has_duplicate_set(arr)
    t3 = time.perf_counter() - start

    print(f"{size:<10} {t1:<15.6f} {t2:<15.6f} {t3:<15.6f}")

print("\nAll three give the SAME answer.")
print("But O(n) is dramatically faster at scale.")
print("At n=1,000,000: O(n^2) would take HOURS, O(n) takes milliseconds.")
````

````python
# DSA in data engineering -- concrete examples
print("DSA IN DATA ENGINEERING -- Real Examples")
print("=" * 60)

# Example 1: GROUP BY uses a hash map
print("\n1. GROUP BY = Hash Map")
orders = [
    {"customer": "Alice", "amount": 100},
    {"customer": "Bob", "amount": 200},
    {"customer": "Alice", "amount": 150},
    {"customer": "Charlie", "amount": 300},
    {"customer": "Bob", "amount": 75},
]

# This is what SQL GROUP BY does internally:
revenue_by_customer = {}  # Hash map: customer -> total
for order in orders:
    customer = order["customer"]
    revenue_by_customer[customer] = revenue_by_customer.get(customer, 0) + order["amount"]

print(f"   Revenue by customer: {revenue_by_customer}")
print(f"   Time complexity: O(n) -- one pass through the data")

# Example 2: Deduplication uses a hash set
print("\n2. DEDUPLICATION = Hash Set")
raw_ids = [1, 2, 3, 2, 4, 1, 5, 3, 6]
seen = set()
unique_ids = []
for id in raw_ids:
    if id not in seen:
        unique_ids.append(id)
        seen.add(id)
print(f"   Raw IDs: {raw_ids}")
print(f"   Unique IDs: {unique_ids}")
print(f"   Time complexity: O(n) -- set lookup is O(1)")

# Example 3: Top-N uses a heap
import heapq
print("\n3. TOP-N CUSTOMERS = Heap (Priority Queue)")
customers = [
    ("Alice", 5000), ("Bob", 12000), ("Charlie", 3000),
    ("Diana", 8000), ("Eve", 15000), ("Frank", 6000),
]
# Get top 3 by revenue -- heap is more efficient than sorting all
top_3 = heapq.nlargest(3, customers, key=lambda x: x[1])
print(f"   All customers: {customers}")
print(f"   Top 3: {top_3}")
print(f"   Time complexity: O(n log k) -- much better than O(n log n) sort")

# Example 4: Pipeline ordering uses topological sort
print("\n4. PIPELINE ORDERING = Topological Sort (Graph)")
dependencies = {
    "ingest_orders": [],
    "ingest_customers": [],
    "transform_silver": ["ingest_orders", "ingest_customers"],
    "build_gold": ["transform_silver"],
    "quality_checks": ["build_gold"],
    "send_report": ["quality_checks"],
}

def topological_sort(deps):
    visited = set()
    order = []
    def visit(node):
        if node in visited:
            return
        visited.add(node)
        for dep in deps.get(node, []):
            visit(dep)
        order.append(node)
    for node in deps:
        visit(node)
    return order

execution_order = topological_sort(dependencies)
print(f"   Execution order: {execution_order}")
print(f"   This is exactly what Airflow/Lakeflow Jobs does internally!")
````

# Section 5: How to Approach a DSA Problem

## The UMPIRE Framework

When you see a problem, follow these steps:

```
U -- Understand the problem
     Read carefully. What are the inputs? What's the expected output?
     Ask clarifying questions. Don't assume.

M -- Match to a pattern
     Does this look like a search problem? A sorting problem?
     Have I seen something similar before?

P -- Plan your approach
     Think before coding. What data structure? What algorithm?
     What's the time/space complexity?

I -- Implement
     Write clean code. Use meaningful variable names.
     Handle edge cases.

R -- Review
     Test with the given examples.
     Test edge cases: empty input, single element, duplicates.

E -- Evaluate
     What's the time complexity? Space complexity?
     Can you do better?
```

## Common Problem Patterns

| Pattern | When to Use | Example |
|---------|------------|---------|
| **Two Pointers** | Sorted array, find pair | Two Sum in sorted array |
| **Sliding Window** | Subarray/substring problems | Max sum subarray of size k |
| **Hash Map** | Count frequencies, find pairs | Two Sum, Group Anagrams |
| **Binary Search** | Sorted data, find element | Search in rotated array |
| **BFS** | Shortest path, level-by-level | Shortest path in grid |
| **DFS** | Explore all paths, backtracking | All permutations |
| **Dynamic Programming** | Overlapping subproblems | Fibonacci, Knapsack |
| **Greedy** | Locally optimal = globally optimal | Activity selection |

## The Most Important Advice

> **Don't try to memorize solutions. Learn the PATTERNS.**
>
> There are ~15 core patterns. Once you recognize them,
> you can solve hundreds of problems.

Start with easy problems. Get comfortable. Then move to medium.
Hard problems are for FAANG -- most DE roles don't need them.

````python
# Problem-solving demonstration: Two Sum
print("PROBLEM-SOLVING DEMO: Two Sum")
print("=" * 60)
print("""
Problem: Given an array of integers and a target sum,
         return the indices of two numbers that add up to the target.

Example:
  Input:  nums = [2, 7, 11, 15], target = 9
  Output: [0, 1]  (because nums[0] + nums[1] = 2 + 7 = 9)
""")

# Step 1: Understand
print("STEP 1: UNDERSTAND")
print("  - Input: list of integers + target integer")
print("  - Output: indices of two numbers that sum to target")
print("  - Assume exactly one solution exists")
print("  - Can't use same element twice")

# Step 2: Match to pattern
print("\nSTEP 2: MATCH TO PATTERN")
print("  - Need to find a PAIR that sums to target")
print("  - For each number x, we need to find (target - x)")
print("  - Hash Map pattern: store seen numbers for O(1) lookup")

# Step 3: Plan
print("\nSTEP 3: PLAN")
print("  - Use a hash map: {number -> index}")
print("  - For each number, check if (target - number) is in the map")
print("  - If yes: return [map[complement], current_index]")
print("  - If no: add current number to map")
print("  - Time: O(n), Space: O(n)")

# Step 4: Implement
print("\nSTEP 4: IMPLEMENT")

def two_sum(nums, target):
    seen = {}  # {number: index}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

# Step 5: Review
print("\nSTEP 5: REVIEW")
test_cases = [
    ([2, 7, 11, 15], 9, [0, 1]),
    ([3, 2, 4], 6, [1, 2]),
    ([3, 3], 6, [0, 1]),
]
for nums, target, expected in test_cases:
    result = two_sum(nums, target)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] two_sum({nums}, {target}) = {result} (expected {expected})")

# Step 6: Evaluate
print("\nSTEP 6: EVALUATE")
print("  Time complexity: O(n) -- one pass through the array")
print("  Space complexity: O(n) -- hash map stores up to n elements")
print("  Could we do O(1) space? Yes, with sorting + two pointers, but O(n log n) time")
print("  Trade-off: O(n) time + O(n) space vs O(n log n) time + O(1) space")
````

# Section 7: Your First Practice Problem

## Problem: FizzBuzz (Classic Warm-Up)

Print numbers 1 to n, but:
- For multiples of 3, print "Fizz"
- For multiples of 5, print "Buzz"
- For multiples of both 3 and 5, print "FizzBuzz"

This is a classic interview warm-up. It tests:
- Basic logic
- Modulo operator
- Order of conditions (FizzBuzz before Fizz/Buzz)

````python
# ============================================
# YOUR TURN: Implement FizzBuzz
# ============================================
# Write a function fizzbuzz(n) that returns a list of strings:
# - "FizzBuzz" for multiples of both 3 and 5
# - "Fizz" for multiples of 3 only
# - "Buzz" for multiples of 5 only
# - The number as a string otherwise
#
# Example: fizzbuzz(15) should return:
# ['1','2','Fizz','4','Buzz','Fizz','7','8','Fizz','Buzz',
#  '11','Fizz','13','14','FizzBuzz']
#
# Write your solution below:
````

````python
# ============================================
# SOLUTION: FizzBuzz
# ============================================
def fizzbuzz(n):
    result = []
    for i in range(1, n + 1):
        if i % 15 == 0:      # Check 15 FIRST (both 3 and 5)
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 == 0:
            result.append("Buzz")
        else:
            result.append(str(i))
    return result

# Test
output = fizzbuzz(20)
print("FizzBuzz(20):")
for i, val in enumerate(output, 1):
    print(f"  {i:2d}: {val}")

# Verify
assert fizzbuzz(15)[-1] == "FizzBuzz", "15 should be FizzBuzz"
assert fizzbuzz(3)[2] == "Fizz", "3 should be Fizz"
assert fizzbuzz(5)[4] == "Buzz", "5 should be Buzz"
print("\nAll assertions passed!")

print("\nKey insight: Check i % 15 BEFORE i % 3 and i % 5")
print("If you check i % 3 first, 15 would print 'Fizz' instead of 'FizzBuzz'")
````
