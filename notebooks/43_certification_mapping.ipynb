## Practice Questions

**Q1:** What does `trigger(availableNow=True)` do in Structured Streaming?
**A:** Processes all available data in multiple micro-batches, then stops. Used for scheduled batch jobs.

**Q2:** What is the difference between `expect` and `expect_or_drop` in DLT?
**A:** `expect` logs violations but keeps the row. `expect_or_drop` removes rows that violate the expectation.

**Q3:** What command compacts small files in a Delta table?
**A:** `OPTIMIZE table_name` -- compacts small files into larger ones for better read performance.

**Q4:** What is the minimum retention period for VACUUM?
**A:** 168 hours (7 days) by default. Can be overridden but not recommended below 24 hours.

**Q5:** What does `cloudFiles.schemaEvolutionMode = "rescue"` do?
**A:** Puts any data that doesn't match the schema into a `_rescued_data` column as JSON, preventing data loss.

# 📗 Section 2: Practice Questions

## Sample Questions (Associate Level)

**Q1:** What is the primary benefit of Delta Lake over plain Parquet files?
- A) Faster read performance
- B) ACID transactions and time travel
- C) Smaller file sizes
- D) Better compression

**Answer: B** — Delta Lake adds ACID transactions, time travel, schema enforcement,
and MERGE support on top of Parquet files.

**Q2:** Which Auto Loader option specifies the format of source files?
- A) `cloudFiles.sourceFormat`
- B) `cloudFiles.format`
- C) `cloudFiles.fileFormat`
- D) `cloudFiles.type`

**Answer: B** — `cloudFiles.format` specifies the source file format (json, csv, parquet, etc.)
